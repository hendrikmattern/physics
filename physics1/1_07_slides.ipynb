## Density

- Density is mass per volume measure in $\text{kg/m}^3$
  $$
  \rho = \frac{m}{V}
  $$

- Temperature dependence:
  $$
  \rho(T) = \frac{\rho_0}{1 + \beta (T - T_0)}
  $$

- **Gases**: strong dependence on $T$ and $p$; **solids/liquids**: weak dependence
- **Water anomaly**: maximum density at $4^\circ\text{C}$

## Pressure - Scalar Definition

- Pressure = normal force $F_\perp$ over contact area $A$:
  $$
  P = \frac{F_\perp}{A}
  $$
- Pressure is a **scalar**
- Units:
  - SI: $1~\text{Pa} = 1~\text{N/m}^2$
  - $1~\text{atm} \approx 1.013\times10^5~\text{Pa}$  
  - $1~\text{bar} = 10^5~\text{Pa}$  
  - $1~\text{mmHg} = 133~\text{Pa}$  
  - $1~\text{psi} = 6895~\text{Pa}$

## Pressure - Differential Form

- Introduce the **area vector**:
  $$
  \mathrm{d}\vec A = \hat n\,\mathrm{d}A
  $$
  - magnitude = area $\mathrm{d}A$  
  - direction = outward surface normal $\hat n$
- **Force** on the surface element:
  $$
  \mathrm{d}\vec F = -p\,\mathrm{d}\vec A
  $$
- **Minus sign**: Convention that force is considered towards the surface element, while the normal vector points outward

## Pressure at a Point in a Fluid

`mf05 - Schweredruckmessung`

- In a fluid at rest, pressure is **the same in all directions**
- Pressure in static fluids does **not** depend on orientation or container shape
- If pressure were not isotropic, there would be a **pressure difference** $\rightarrow$ a net force would act $\rightarrow$ the fluid would start to flow
- Fluids exert forces **perpendicular** to surfaces.



## Pressure & Depth – Derivation

(Take $y$ as vertical coordinate, positive upward.)

- thin fluid layer of thickness $\mathrm{d}y$, area $A$, and mass $\rho A\,\mathrm{d}y$
- Pressure forces & weight:
  - **upward** (from the bottom surface): $(P + \mathrm{d}P)A$
  - **downward** (from the top surface): $PA$
  - **downward** weight: $\rho g A\,\mathrm{d}y$

- Force balance in $y$ (no acceleration):
  $$
  (P + \mathrm{d}P)A \;-\; PA \;-\; \rho g\,A\,\mathrm{d}y = 0
  \quad\Rightarrow\quad
  \mathrm{d}P\,A - \rho g\,A\,\mathrm{d}y = 0
  \quad\Rightarrow\quad
  \boxed{\mathrm{d}P = \rho g\,\mathrm{d}y}

  $$

- Since pressure increases **downward** while $y$ increases **upward**, we have
  $$
  \boxed{\frac{\mathrm{d}P}{\mathrm{d}y} = -\,\rho g}
  $$


## Pressure \& Depth - Absolute Pressure

- At the fluid surface ($y = 0$), pressure is $P_0$ (e.g.  atmospheric pressure)

- Integrate from $0$ to depth $h$:
  $$
  \int_{P_0}^P \mathrm{d}P = \int_0^h  \rho g\,\mathrm{d}y
  $$
  $$
  P - P_0 = \rho g h
  $$
  $$
  P = P_0 + \rho g h
  $$

- Result: pressure in a uniform-density fluid **increases linearly with depth**

## Pressure Differences & Hydrostatic Paradox (cont')

- Pressure difference in a uniform fluid:
  $$
  \Delta P = \rho g (h_2 - h_1)
  $$
- Pressure depends only on **depth**, not container shape or volume
- **Hydrostatic paradox:** containers of different shape have the **same pressure** at the same depth
- **Reason**: Pressure is set by the **fluid column above the point**, not by total fluid amount


## Atmospheric Pressure

- Air has mass $\rightarrow$ its weight creates **atmospheric pressure** on all surfaces  
- Pressure at a point is the weight of the air column above:
  $$
  P = \int_h^\infty \rho g\,\mathrm{d}h
  $$
- At sea level:
  $$
  P_{\text{atm}} \approx 1.013\times10^5~\text{Pa} \approx 1~\text{atm}
  $$
- Equivalent units (single line): **101.3 kPa = 760 mmHg = 14.7 psi = 1.013 bar**
- Atmospheric pressure corresponds to a **10.3 m water column**


## Atmospheric Pressure vs. Altitude

- Higher altitude $\rightarrow$ **less air above** $\rightarrow$ lower pressure  
- For small height changes, density $\approx$ constant, thus, linear approximation gives $P = P_0 - \rho g h$  

- For larger height differences, density varies (ideal gas assumed): $\rho = \frac{M P}{R T}$
- Substituting into hydrostatic relation:
  $$
  \frac{\mathrm{d}P}{\mathrm{d}h} = -\rho g \Rightarrow \; \frac{\mathrm{d}P}{\mathrm{d}h} = -\frac{M g}{R T} P
  $$
- Solve by separation:
  $$
  P = P_0\, e^{-h/h_0}
  $$
- **Scale/characteristic height**:
  $$
  h_0 = \frac{R T}{M g}
  $$
  The height over which pressure drops by a factor of $e$ (Earth: $h_0 \approx 8.4~\mathrm{km}$ at 20°C)


## An Ode to Ordinary Differential Equations (ODE)

- Hydrostatic relation is a **differential equation**:
  $$
  \frac{\mathrm{d}P}{\mathrm{d}h} = -\rho g
  $$
- Its solution depends on the model for density:
  - **Constant $\rho$** → linear law  
    $P = P_0 + \rho g h$
  - **Ideal-gas atmosphere** → exponential law  
    $P = P_0\, e^{-h/h_0}$
- This ODE framework allows us to model any atmosphere  
  (e.g., variable temperature, variable composition)

## Absolute, Gauge, Differential Pressure & Vacuum

- **Absolute pressure:** measured relative to **vacuum** ($P = 0$)
- **Atmospheric pressure:** ambient air pressure $P_{\text{atm}}$
- **Gauge pressure:** relative to atmosphere  
  $$
  P_g = P_{\text{abs}} - P_{\text{atm}}
  $$
  - $P_g>0$: above atmospheric  
  - $P_g<0$: **partial vacuum**
- **Differential pressure:**  
  $$
  \Delta P = P_2 - P_1
  $$


## Measuring Pressure with Fluid Columns (Manometers)

A manometer measures the pressure difference between **system 1** and **system 2**:
$$
P_1 - P_2 = \rho g (h_2 - h_1)
$$

- If **system 2 = atmosphere**, this reduces to:
  $$
  P_{\text{system}} = P_{\text{atm}} - \rho g h
  $$

- **positive \& negative pressure differences can be detected**

- Invented by Evangelista Torricelli (1608–1647) to measure **absolute** atmospheric pressure
- A tube filled with mercury is inverted into a reservoir $\rightarrow$ a small **vacuum** forms at the top (no trapped air)

  $$
  P_{\text{atm}} = \rho_{\text{Hg}} g h

## Pascal's Principle

`mf02 - Allseitge Druckausbreitung`

> **Pascal’s Principle:**  
> A change in pressure applied to an enclosed fluid is transmitted undiminished to every point of the fluid and to the walls of the container.

This means that an applied pressure change has the same value everywhere in the fluid:
$$
\Delta P = \frac{\Delta F}{A}
$$

## Hydraulic Press - Theory

- **Pascal’s principle**:
  $$
  F_2 = P A_2 = \frac{A_2}{A_1} F_1
  $$
- **Mechanical advantage**: $\displaystyle \frac{F_2}{F_1} = \frac{A_2}{A_1}$

- **Energy conserved**: large force ↔ small displacement  
  $$
  W_1 = P A_1 d_1 = P A_2 d_2 = W_2, \; \Rightarrow \; F_1 d_1 = F_2 d_2
  $$

## Buoyant Force - Origin

- Pressure increases with depth: $P = \rho g h$
- Forces on top and bottom of a submerged object (area $A$):
  $$
  F_{\text{top}} = \rho g h_t A, \qquad
  F_{\text{bottom}} = \rho g h_b A
  $$
- Net upward (buoyant) force:
  $$
  F_B = F_{\text{bottom}} - F_{\text{top}}
      = \rho g (h_b - h_t)A
      = \rho g V_{\text{disp}}
  $$

- Apparent weight:
  $$
  W_{\text{app}} = W - F_B
  $$

## Buoyancy - Floating condition

`mf04 - Flaschenteufel`

- Net upward (buoyant) force: $F_B = \rho_{\text{fluid}} g V_{\text{disp}}$

- Apparent weight: $W_{\text{app}} = W - F_B = \rho_{\text{obj}} g V_{\text{obj}} - \rho_{\text{fluid}} g V_{\text{disp}}$

- Floating condition ($W_{\text{app}} = 0 \; \leftrightarrow \; W = F_B$):  
  $$
  \frac{V_{\text{sub}}}{V_{\text{obj}}}
  = \frac{\rho_{\text{obj}}}{\rho_{\text{fluid}}}
  $$

## Surface Tension - Molecular Perspective

- **Cohesion**: attraction between molecules of the same substance
- **Adhesion**: attraction between molecules of different substances
- **Bulk molecules**: many neighbors → many cohesive bonds → low energy  
- **Surface molecules**: fewer neighbors → fewer bonds → higher energy (costs work to create surface)

## Definition of Specific Surface Energy & Surface Tension

- Enlarging a liquid surface by $\Delta A$ requires work $\Delta W$.  
  **Specific surface energy:**  
  $$\varepsilon = \frac{\Delta W}{\Delta A}$$

- A liquid surface pulls tangentially along any line in the interface.  
  **Surface tension:**  
  $$\gamma = \frac{F}{L}$$

- In static equilibrium:  
  $$\gamma = \varepsilon$$

**Units:** N/m = J/m²  (since J/m² = N·m/m² = N/m)


## Capillarity: Cohesion vs Adhesion

`mf17 - Kapillarität`

- **Adhesion > Cohesion**: $\rightarrow$ liquid wets the tube $\rightarrow$ concave meniscus  $\rightarrow$ liquid **rises**  

- **Cohesion > Adhesion**: $\rightarrow$ liquid does not wet the tube $\rightarrow$ convex meniscus $\rightarrow$ liquid **falls** (depression)  

- Capillary height is **inversely proportional** to tube radius

**If connected, what bubble grows?**

$\rightarrow$ pressure difference large the smaller the droplet

### Excess Pressure in a Spherical Droplet (Surface–Energy Derivation)

Small radial increase: $r \to r + \Delta r$

**Work by pressure**
$$
\Delta V = 4\pi r^2 \Delta r,
\qquad
\Delta W = p\,\Delta V = p(4\pi r^2 \Delta r)
$$

**Increase in surface energy**
$$
E = \varepsilon (4\pi r^2)
$$
$$
\Delta E = 4\pi\varepsilon\big[(r+\Delta r)^2 - r^2\big]
       \approx 8\pi r\varepsilon\,\Delta r
$$

**Energy balance**
$$
\Delta W = \Delta E
$$
$$
p(4\pi r^2 \Delta r) = 8\pi r\varepsilon\,\Delta r
$$

**Result**
$$
p = \frac{2\varepsilon}{r}
\qquad (\gamma = \varepsilon)
\qquad\Rightarrow\qquad
\Delta P = \frac{2\gamma}{r}
$$


### Soap Bubble (two surfaces)

Twice the surface energy:
$$
\Delta E_{\text{bubble}} = 2 \times (8\pi r\varepsilon\,\Delta r)
$$

Result:
$$
\Delta P = \frac{4\gamma}{r}
$$

A soap bubble has **twice** the excess pressure of a droplet of the same radius.
