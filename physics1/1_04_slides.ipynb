## Definition of Work
- In physics, work has a strict quantitative meaning (not “effort” as in everyday language)
- **Work describes energy transfer caused by a force acting on an object along the direction of displacement**

## Work for Straight-Line Motion under a Constant Force

- Work is defined by the dot product of $\vec{F}$ and the displacement $\vec{d}$:
$$
W = \vec F \cdot \vec d
$$

- Dot product $\rightarrow$ work is a **scalar**
- Inputs: force vector + displacement vector
- Units:
$$
1~\text{J} = 1~\text{N}\cdot\text{m} = 1~\text{kg}\cdot\text{m}^2/\text{s}^2
$$

## Work for Straight-Line Motion under a Constant Force (cont')

- With $\theta$ as the angle between $\vec{F}$ \& $\vec{d}$
$$
W = \vec F \cdot \vec d = F d \cos\theta
$$

- Projection interpretation:
$$
\vec F \cdot \vec d = F_\parallel d
$$
$\rightarrow$ Only component of force parallel to displacement does work.

## Work and constant Force: Solutions

| Example                        | Situation / Key Idea                                      | Work Result          |
|-------------------------------|------------------------------------------------------------|----------------------|
| Pushing against a wall        | Force applied, but **displacement = 0**                    | **no work**          |
| Lifting an object up             | Force ↑, displacement ↑                                    | **positive work**    |
| Lowering an object slowly     | Force ↑, displacement ↓                                    | **negative work**    |
| Carrying an object            | Force ↑, displacement → (horizontal), angle = $90^\circ$   | **no work**          |
| Pulling a box at an angle     | Only horizontal component $F_\parallel$ contributes        | **positive work**    |
| Sliding a box with friction   | Applied force → positive; friction → negative              | **net work = sum of works** |
| Circular motion               | Centripetal force $\perp$ velocity                         | **no work**          |



## Work and constant Force: Summary

- Work from constant force:
$$
W = \vec F \vec d = F d \cos\theta
$$

- Sign from angle:
  - $0^\circ \le \theta < 90^\circ$ $\rightarrow$ **positive work**
  - $\theta = 90^\circ$ $\rightarrow$ **zero work**
  - $90^\circ < \theta \le 180^\circ$ $\rightarrow$ **negative work**

- Only the **parallel component** of force transfers energy

- Net work: $W_{\text{net}} = \sum_i W_i = \sum_i \vec F_i \vec d = \vec F_{\text{net}}\, \vec d$


## Work Done by a Variable Force

- Force may vary in magnitude and/or direction
- Split path into infinitesimal segments along the position vector $\vec r$:
$$
\mathrm dW = \vec F(\vec r)\cdot \mathrm d\vec r
$$

- Total work = sum of all infinitesimal contributions:
$$
W = \int \mathrm dW = \int \vec F(\vec r)\cdot \mathrm d\vec r
$$

- Dot product ensures only the **parallel component** contributes


## Work as a Line Integral

- General form between $\vec r_1$ and $\vec r_2$: $W = \int_{\vec r_1}^{\vec r_2} \vec F(\vec r)\cdot \mathrm d\vec r$ $\Rightarrow$ **signed area under the $F(\vec r)$ curve**

- 1D case: 
$$W = \int_{x_1}^{x_2} F(x)\,\mathrm dx$$
    
- 3D case:
$$
W = \int_{x_1}^{x_2} F_x\,\mathrm dx + \int_{y_1}^{y_2} F_y\,\mathrm dy + \int_{z_1}^{z_2} F_z\,\mathrm dz
$$

## Hooke’s Law: Spring Force
`mi17 - Feder` 

- Ideal spring force with spring constant (stiffness) $k$ \& displacement from equilibrium $x$:
$$
F_S = -k x
$$

- Negative sign $\rightarrow$ restoring force (points opposite to $x$)
- Force increases **linearly** with displacement  
  → straight line in the $F$–$x$ diagram

## Work Done by a Spring

- Work done **by the spring** (from $x_1$ to $x_2$):
$$
W_S = \int_{x_1}^{x_2}  F_S\,\mathrm dx = \int_{x_1}^{x_2} (-k x)\,\mathrm dx
     = -\tfrac12 k (x_2^2 - x_1^2)
$$

- Work done **on the spring** by external force to stretch from $0$ to $x$:
$$
F_{\text{ext}} = +k x
$$
$$
W_{\text{ext}} = \int_{0}^{x} (+k x)\,\mathrm dx = \tfrac12 k x^2
$$
- $\rightarrow$ equal to **area under the $F$–$x$ graph** (triangle)


## Deriving Kinetic Energy from Work

`md36 - Schuss`

- Infinitesimal work:  
  $$ \mathrm dW = \vec F_{\text{net}} \cdot \mathrm d\vec l = m\vec a \cdot \mathrm d\vec l  $$

- With $ \vec a = \dfrac{d\vec v}{dt} $ and $ \mathrm d\vec l = \vec v\,\mathrm dt $:  
  $$ \mathrm dW = m\,\dfrac{d\vec v}{dt}\cdot \vec v\,\mathrm dt = m\,\vec v\cdot\mathrm d\vec v $$

- Integrate from $ \vec v_1 $ to $ \vec v_2 $:  
  $$ W_{\text{net}} = \int_{\vec v_1}^{\vec v_2} m\,\vec v\cdot\mathrm d\vec v = \tfrac12 m v_2^2 - \tfrac12 m v_1^2 $$

- Define work–energy relation \& kinetic energy:  
  $$ W_{\text{net}} = \Delta K, \quad K = \tfrac12 m v^2$$


## Energy 

> Energy is a scalar quantity that measures the ability of a system to perform work or to produce changes in the system or its environment.

$\rightarrow$ Energy also measure in joule [J]

$\rightarrow$ Several forms of energy, e.g. kinetic energy describing motion, or potential energy representing capacity to do work due to position/configuration

## Deriving Potential Energy from Work

- Total work from A to B:  
  $$ W_{AB} = \int \vec F(\vec r)\cdot\mathrm d\vec l $$

- For **conservative forces** (gravity, spring), **work depends only on start/end points**

- Define potential energy change:  
  $$ \Delta U = U_B - U_A = -W_{AB} = -\int \vec F(\vec r)\cdot\mathrm d\vec l $$



## Gravitational Potential Energy

`md25 - Anheben Kette`

- Gravitational force (uniform field) wih $\hat y$ pointing upwards:
$$
\vec F_G = -mg\,\hat y
$$

- Work done by gravity on the object:
$$
W_G = \int_{y_1}^{y_2} (-mg)\,\mathrm dy
     = -mg (y_2 - y_1)
$$

- Define potential energy with $U=0$ at $y=0$:
$$
U_{\text{grav}} = -W_G =  mgy
$$


## Elastic Potential Energy (Spring)

- Spring force obeying Hooke’s law:
$$
F_S = -k x
$$

- Work done by the spring on the object:
$$
W_S = \int_{x_1}^{x_2} (-k x)\,\mathrm dx
     = -\tfrac12 k (x_2^2 - x_1^2)
$$

- Define potential energy with $U(x=0)=0$:
$$
U_{\text{elastic}} = -W_S = \tfrac12 k x^2
$$

## Reference Point & Force–Potential Relation

## Conservative vs. Nonconservative Forces

- **Conservative forces**:
  - **Work independent of path**, e.g. gravity, springs, electrostatic force  
  - Closed-path work: $\oint \vec F\cdot\mathrm d\vec l = 0$  
- **Nonconservative forces**:
  - **Work depends on path**, e.g. friction, air resistance, drag  
  - Mechanical energy not fully recoverable

## Conservation of Energy

`md35 - Vase`

`md29 - Looping` 

> **Law of Conservation of Energy:**  
> The total energy of an isolated system remains constant.  
> Energy may be transferred or transformed, but cannot be created or destroyed.

- Total energy includes all forms:  
  mechanical, thermal, chemical, electrical, nuclear, radiant

- Dissipative forces (friction, drag):  
  convert mechanical energy $\rightarrow$ thermal energy

- Energy balance (isolated system):
  $$K + U + E_{\text{other}} = \text{const} \; \Leftrightarrow \; \Delta(K + U + E_{\text{other}}) = 0$$

## Power & Efficiency

- **Power** is the **rate of doing work**:
  $$P_{\text{avg}} = \dfrac{W}{\Delta t}, \quad \& \quad P = \dfrac{\mathrm dW}{\mathrm dt} = \frac{\vec{F} \mathrm d\vec{l}}{\mathrm dt} = \vec F \cdot \vec v$$

- **Units**: watt (W = J/s)

- **Efficiency** (useful output vs. input):
  $\eta = \dfrac{E_{\text{useful}}}{E_{\text{input}}}$

- Express as decimal or percentage;  
  ideal $\eta = 1$ not achievable due to losses
