### Refraction of Sound Waves

Sound waves refract when they travel through regions of air with different temperatures or densities, since the speed of sound depends on these properties. In the atmosphere, temperature gradients cause sound waves to bend. When the air near the ground is warmer than the air above, sound tends to bend upward, while cooler air near the ground causes sound waves to bend downward. This explains why distant sounds are often heard more clearly at night, when cooler air near the surface refracts sound back toward the ground.

### Pressure Variation in Longitudinal Waves

In longitudinal waves, the oscillatory particle motion leads to variations in pressure within the medium. The pressure fluctuation associated with the wave can be written as
$$
\Delta P(x,t) = B k s_m \sin(kx - \omega t),
$$
where $B$ is the bulk modulus of the medium, which characterizes its resistance to compression, and $s_m$ is the displacement amplitude. The pressure variation is therefore **90 degrees out of phase** with the particle displacement, meaning that the pressure is maximum where the displacement passes through zero.

### Energy Transport in Sound Waves

In longitudinal sound waves, energy is transported through alternating compressions and rarefactions of the medium. The average intensity of a sound wave is given by
$$
I = \frac{1}{2} \rho v \omega^2 s_m^2,
$$
where $\rho$ is the density of the medium, $v$ is the speed of sound in that medium, and $s_m$ is the amplitude of the particle displacement.

### Sound Intensity and Loudness

The range of sound intensities detectable by the human ear extends from the **threshold of hearing** to the **threshold of pain** and spans many orders of magnitude. Because of this wide range, sound intensity is expressed on a logarithmic scale known as the **decibel scale**. The sound level is defined as
$$
\beta = 10 \log_{10}\!\left(\frac{I}{I_0}\right)\,\text{dB},
$$
where $I_0$ is the reference intensity corresponding to the threshold of hearing.

### Derivation of Wave Equation from a Tensioned String

For a stretched string under tension $T$ and with linear mass density $\mu$, the wave equation can be derived directly from Newton’s second law. The net vertical force on a small string element arises from the difference in tension at its ends and is proportional to the second spatial derivative of the displacement. Applying Newton’s second law to the vertical motion of the element leads to
$$
\mu \frac{\partial^2 y}{\partial t^2} = T \frac{\partial^2 y}{\partial x^2}.
$$
Rearranging this expression yields
$$
\frac{\partial^2 y}{\partial x^2} = \frac{\mu}{T}\frac{\partial^2 y}{\partial t^2}.
$$
The wave speed on a stretched string is therefore
$$
v = \sqrt{\frac{T}{\mu}},
$$
and substituting this expression into the equation above recovers the standard form of the wave equation.

### Interference in Sound Waves

When two sound waves of the same frequency overlap, the perceived loudness depends on their relative phase. Regions of constructive interference correspond to increased intensity and sound louder, while regions of destructive interference correspond to reduced intensity and sound quieter. If the two sound waves have slightly different frequencies, the interference pattern changes with time, leading to an audible modulation of loudness.

### Beats and Interference in Time

When two waves with slightly different frequencies interfere, their superposition produces a rapidly oscillating motion whose amplitude varies slowly in time. This phenomenon is described by
$$
y(t) = 2A \cos\!\left[\pi(f_1 - f_2)t\right]\sin\!\left[2\pi\left(\frac{f_1 + f_2}{2}\right)t\right],
$$
and the amplitude modulation occurs at the **beat frequency**
$$
f_{\text{beat}} = |f_1 - f_2|.
$$
The periodic increase and decrease in amplitude is perceived as beats, a common effect in acoustics.

### Standing Waves in Air Columns

Standing sound waves can also form in columns of air, such as in organ pipes and resonance tubes. If the tube is open at both ends, both ends behave as displacement antinodes, and the allowed wavelengths and frequencies are the same as for a string fixed at both ends,
$$
\lambda_n = \frac{2L}{n}, \qquad f_n = \frac{n v}{2L}, \qquad n = 1,2,3,\dots.

$$
\lambda_n = \frac{4L}{n}, \qquad f_n = \frac{n v}{4L}, \qquad n = 1,3,5,\dots.

### Standing Waves on a String and Resonance

For a string of length $L$ fixed at both ends, the displacement must vanish at $x=0$ and $x=L$. Applying this boundary condition to the standing-wave form requires
$$
\sin(kL)=0,
$$
which implies
$$
kL = n\pi, \qquad n = 1,2,3,\dots.
$$
The allowed wavelengths and frequencies are therefore
$$
\lambda_n = \frac{2L}{n}, \qquad
f_n = \frac{n v}{2L}.
$$
These discrete frequencies are the **natural frequencies** or **harmonics** of the string. Each allowed standing-wave pattern is called a **mode of vibration**. The lowest-frequency mode is the **fundamental**, while higher modes are overtones.

**Resonance** occurs when a system is driven by an external periodic force whose frequency matches one of its natural frequencies. Under this condition, energy transfer is highly efficient and the amplitude of oscillation increases significantly. In real systems, the amplitude is ultimately limited by damping and energy losses.

In a standing wave, kinetic energy is largest near nodes where the displacement changes most rapidly, while potential energy is largest near antinodes where the displacement is greatest. While resonance is essential for sound production and signal amplification, excessive resonance can lead to structural damage if oscillations grow too large.

## 1.11.1 Characteristics of Sound

Sound is a **mechanical wave** produced by the vibration of matter and transmitted through a material medium by successive **compressions** and **rarefactions** of its particles. Because sound requires a medium for propagation, it cannot travel through a vacuum. The physical properties of sound waves determine how sound is perceived in terms of pitch, loudness, and quality.

### Nature of Sound Waves

Sound waves in gases and liquids are **longitudinal**, meaning that particles of the medium oscillate **parallel** to the direction of wave propagation. In solids, sound waves may be either longitudinal or transverse, depending on the type of deformation involved. Sound is generated by vibrating sources such as vocal cords, tuning forks, or loudspeakers, and it propagates through the surrounding medium as a traveling pattern of pressure variations.

### Regions of Compression and Rarefaction

A **compression** is a region in which particles of the medium are closer together than average and the pressure is higher than equilibrium. A **rarefaction** is a region in which particles are farther apart and the pressure is lower than equilibrium. As compressions and rarefactions move through the medium, they form a traveling longitudinal wave that transfers energy without any net transport of matter.

### Speed of Sound

The **speed of sound** $v$ depends on the medium’s **elastic properties**, which provide the restoring force, and its **inertia**, which resists motion. For a fluid medium such as a gas or liquid, the speed of sound is given by
$$
v = \sqrt{\frac{B}{\rho}},
$$
where $B$ is the bulk modulus of the medium and $\rho$ is its mass density. Sound therefore travels faster in media that are stiffer and less dense, which explains why sound generally propagates fastest in solids, more slowly in liquids, and slowest in gases.

### Dependence of Sound Speed on Temperature

In gases, and especially in air, the speed of sound depends on temperature because both the elasticity and density of the gas vary with temperature. For an ideal gas, the speed of sound is
$$
v = \sqrt{\frac{\gamma R T}{M}},
$$
where $\gamma$ is the ratio of specific heats, $R$ is the universal gas constant, $T$ is the absolute temperature, and $M$ is the molar mass of the gas. Over ordinary temperature ranges in air, the speed of sound increases approximately linearly with temperature, rising by about $0.6~\text{m/s}$ for each degree Celsius increase.

### Frequency, Wavelength, and Pitch

Sound waves satisfy the general wave relation
$$
v = f\lambda,
$$
where $f$ is the frequency, $\lambda$ is the wavelength, and $v$ is the speed of sound in the medium. The **frequency** determines the **pitch** of the sound, which is how high or low it is perceived by the human ear. Frequency is fixed by the source of the sound and does not change when the sound enters a different medium, although the wavelength does change if the speed changes.

### Audible, Infrasonic, and Ultrasonic Sound

Human hearing is limited to a finite range of frequencies. Sounds with frequencies below the audible range are called **infrasonic**, while those above the audible range are called **ultrasonic**. Many natural and technological sources produce sound outside the range of human hearing, and the sensitivity of the ear decreases significantly at both very low and very high frequencies.

### Loudness and Intensity

**Loudness** is the subjective perception of how strong or intense a sound appears to a listener, whereas **intensity** is an objective physical quantity defined as the energy transmitted per unit area per unit time. Although loudness increases with intensity, it also depends on frequency and on the sensitivity of the human ear. The average intensity of a sound wave is proportional to the **square of the amplitude** of the pressure variation.

### Characteristics of Sound Waves

A sound wave is fully characterized by four fundamental quantities. The **amplitude** determines the loudness, the **frequency** determines the pitch, the **waveform** determines the quality or timbre of the sound, and the **speed** depends on the medium and its temperature. Together, these properties provide a complete physical description of sound.

### Sound as a Pressure Wave

Because sound waves consist of compressions and rarefactions, they can be described in terms of pressure variations about the equilibrium pressure of the medium. A simple representation of a traveling sound wave is
$$
\Delta P(x,t) = \Delta P_{\text{max}} \sin(kx - \omega t),
$$
where $\Delta P_{\text{max}}$ is the amplitude of the pressure variation, $k$ is the wave number, and $\omega$ is the angular frequency. Pressure maxima correspond to compressions, while pressure minima correspond to rarefactions.

### Sound Energy and Intensity Relation

For a plane progressive sound wave, the average intensity is given by
$$
I = \frac{1}{2}\rho v \omega^2 s_m^2,
$$
where $\rho$ is the density of the medium, $v$ is the speed of sound, $s_m$ is the amplitude of particle displacement, and $\omega = 2\pi f$. This relation shows that sound intensity increases with both the square of the amplitude and the square of the frequency.

### Conceptual Notes

Sound is a **mechanical longitudinal wave** that requires a material medium for propagation. Its speed depends on the elasticity and density of the medium and increases with temperature in gases. The frequency of a sound wave determines its pitch and is fixed by the source, while the wavelength adjusts to changes in wave speed. The audible range of human hearing extends from roughly $20~\text{Hz}$ to $20~\text{kHz}$. Intensity is proportional to the square of the amplitude, while loudness is a subjective perceptual response. Sound generally travels fastest in solids, more slowly in liquids, and slowest in gases.

## 1.11.2 Mathematical Representation of Longitudinal Waves

Sound waves in air and other media are **longitudinal waves**, meaning that particles of the medium oscillate **parallel** to the direction of wave propagation. These waves can be described mathematically using particle displacement, particle velocity, and pressure variations as functions of position and time.

### Displacement Representation

Consider a plane sound wave traveling in the positive x-direction. If $s(x,t)$ denotes the displacement of a particle from its equilibrium position, the wave can be written as
$$
s(x,t) = s_m \cos(kx - \omega t),
$$
where $s_m$ is the displacement amplitude, $k = 2\pi/\lambda$ is the wave number, $\omega = 2\pi f$ is the angular frequency, $\lambda$ is the wavelength, $f$ is the frequency, and the wave speed is given by $v = \omega/k$. Each particle of the medium executes simple harmonic motion with amplitude $s_m$ and frequency $f$.

### Particle Velocity and Acceleration

The particle velocity is obtained by differentiating the displacement with respect to time,
$$
v_p(x,t) = \frac{\partial s}{\partial t} = \omega s_m \sin(kx - \omega t).
$$
Differentiating once more yields the particle acceleration,
$$
a(x,t) = \frac{\partial^2 s}{\partial t^2} = -\omega^2 s_m \cos(kx - \omega t) = -\omega^2 s.
$$
This relation shows that the acceleration is proportional to the displacement and opposite in direction, which is the defining characteristic of simple harmonic motion.

### Pressure Variation in a Sound Wave

The oscillatory motion of particles produces small pressure variations about the equilibrium pressure of the medium. The excess pressure $\Delta P(x,t)$ is related to the spatial derivative of the displacement by
$$
\Delta P(x,t) = -B \frac{\partial s}{\partial x},
$$
where $B$ is the bulk modulus of the medium. Substituting the displacement expression gives
$$
\Delta P(x,t) = B k s_m \sin(kx - \omega t).
$$
This result shows that pressure variations are shifted in phase by $90^\circ$ relative to displacement, so pressure is maximum when displacement passes through equilibrium and zero when displacement is maximal.

### Pressure Amplitude

The maximum pressure variation in the sound wave is
$$
\Delta P_{\text{max}} = B k s_m.
$$
Using the relations $v = \sqrt{B/\rho}$ and $\omega = vk$, this can be written as
$$
\Delta P_{\text{max}} = \rho v \omega s_m.
$$
The pressure amplitude therefore depends on the displacement amplitude, the wave frequency, and the physical properties of the medium.

### Phase Relations Between Physical Quantities

In a longitudinal sound wave, particle displacement varies as $\cos(kx - \omega t)$, while particle velocity and pressure variation vary as $\sin(kx - \omega t)$. As a result, particle velocity and pressure are **in phase** with each other and both lead the displacement by $90^\circ$. This phase relationship explains why compressions and rarefactions occur where particle displacement is zero.

### Wave Equation for Sound

Both the displacement field and the pressure field of a sound wave satisfy the one-dimensional wave equation,
$$
\frac{\partial^2 s}{\partial x^2} = \frac{1}{v^2}\frac{\partial^2 s}{\partial t^2}, \qquad
\frac{\partial^2 (\Delta P)}{\partial x^2} = \frac{1}{v^2}\frac{\partial^2 (\Delta P)}{\partial t^2}.
$$
Thus, sound propagation can be described equivalently in terms of particle motion or pressure variations.

### Energy in a Sound Wave

As a sound wave travels, each element of the medium alternately stores kinetic energy due to particle motion and potential energy due to compression and expansion. The average intensity, defined as the power transmitted per unit area, is given by
$$
I = \frac{1}{2}\rho v \omega^2 s_m^2.
$$
This expression shows that sound intensity increases with the square of both the displacement amplitude and the frequency.

### Visualization of Phase Relations

Within one wavelength of a sound wave, displacement varies as a cosine function while pressure varies as a sine function. Points of maximum compression and rarefaction occur where particle displacement is zero, and points of maximum displacement correspond to zero pressure variation. This spatial phase shift underlies the longitudinal nature of sound.

### Pressure and Displacement Summary

Particle displacement is described by $s(x,t) = s_m \cos(kx - \omega t)$, particle velocity by $v_p(x,t) = \omega s_m \sin(kx - \omega t)$, and pressure variation by $\Delta P(x,t) = \rho v \omega s_m \sin(kx - \omega t)$. Particle velocity and pressure are in phase with each other and lead displacement by $90^\circ$, while the sound intensity represents a time-averaged energy flow.

### Conceptual Notes

Sound waves are longitudinal mechanical waves consisting of alternating compressions and rarefactions. Particle displacement and pressure oscillations are out of phase by $90^\circ$, while particle velocity and pressure are in phase. The pressure amplitude is proportional to $\rho v \omega s_m$, and both displacement and pressure satisfy the same wave equation. The average intensity of a sound wave depends on the density of the medium and on the square of the frequency and displacement amplitude.

## 1.11.3 Intensity of Sound and the Decibel Scale

The **intensity** of a sound wave is defined as the rate at which sound energy is transmitted through a unit area perpendicular to the direction of wave propagation. It is an objective, measurable physical quantity that is closely related, but not identical, to the **perceived loudness** of sound. Because the human ear can detect sound intensities spanning many orders of magnitude, sound levels are commonly expressed on a **logarithmic scale** using **decibels (dB)**.

### Definition of Sound Intensity

For a plane progressive sound wave, the average sound intensity is given by
$$
I = \frac{P}{A} = \frac{1}{2}\rho v \omega^2 s_m^2,
$$
where $\rho$ is the density of the medium, $v$ is the speed of sound, $\omega = 2\pi f$ is the angular frequency, and $s_m$ is the amplitude of particle displacement. The factor of one half arises from averaging the oscillatory energy flow over one full cycle. This relation shows that sound intensity increases strongly with both the **square of the amplitude** and the **square of the frequency**.

### Sound Intensity Level and the Decibel Scale

The range of sound intensities detectable by the human ear extends from approximately $10^{-12}~\text{W/m}^2$, known as the threshold of hearing, up to about $1~\text{W/m}^2$, near the threshold of pain. To represent this enormous range conveniently, sound intensity is expressed logarithmically as a **sound intensity level**
$$
\beta = 10 \log_{10}\!\left(\frac{I}{I_0}\right),
$$
where $\beta$ is measured in decibels and $I_0 = 10^{-12}~\text{W/m}^2$ is the reference intensity corresponding to the threshold of hearing. The decibel scale reflects the approximately logarithmic response of the human ear to changes in sound intensity.

### Relative Changes in Sound Level

A change in sound level corresponds to a multiplicative change in intensity according to
$$
\Delta \beta = 10 \log_{10}\!\left(\frac{I_2}{I_1}\right).
$$
An increase of $10~\text{dB}$ corresponds to a tenfold increase in intensity, while an increase of $3~\text{dB}$ corresponds approximately to a doubling of intensity. Perceptually, a $10~\text{dB}$ increase is typically experienced as roughly a doubling of loudness, even though the physical intensity has increased by a factor of ten.

### Typical Sound Levels

Common sound sources span a wide range of intensities and sound levels, from the threshold of hearing at $0~\text{dB}$, through quiet environments and normal conversation around $60~\text{dB}$, to loud traffic, amplified music, and jet engines approaching or exceeding $120~\text{dB}$. The logarithmic nature of the decibel scale allows these vastly different intensities to be represented by relatively small numerical values.

### Combination of Independent Sound Sources

When multiple independent and uncorrelated sound sources act together, their **intensities add**, not their sound levels. For two identical sources each producing intensity $I$, the total intensity is $2I$. The corresponding sound level increases by
$$
\Delta \beta = 10 \log_{10}(2) \approx 3~\text{dB}.
$$
Thus, doubling the number of identical sound sources results in only a modest increase in sound level, even though the total intensity has doubled.

### Dependence of Intensity on Distance

For a point source that radiates sound uniformly in all directions, the sound intensity at a distance $r$ from the source is
$$
I = \frac{P}{4\pi r^2},
$$
where $P$ is the total acoustic power emitted. Sound intensity therefore follows an **inverse-square law** with distance, decreasing as the square of the distance from the source. Doubling the distance from a sound source reduces the intensity by a factor of four, corresponding to a decrease of approximately $6~\text{dB}$.

### Thresholds of Human Hearing and Safety

The **threshold of audibility** corresponds to $0~\text{dB}$, while the **threshold of pain** lies near $120~\text{dB}$. Prolonged exposure to sound levels above approximately $85~\text{dB}$ can cause permanent hearing damage, with risk increasing as both sound level and exposure time increase. These limits are important in occupational safety, environmental noise regulation, and audio engineering.

### Energy, Power, and Intensity

The total acoustic power emitted by a sound source can be related to the measured intensity at a given distance using
$$
P = I \, 4\pi r^2.
$$
This relation allows the intrinsic strength of a sound source to be determined from intensity measurements made at known distances.

### Conceptual Notes

Sound intensity measures the rate of energy flow per unit area and provides an objective description of sound strength. The decibel scale compresses a vast range of intensities into manageable values that align with human perception. An increase of $10~\text{dB}$ corresponds to a tenfold increase in intensity, while a $3~\text{dB}$ increase corresponds to a doubling of intensity. Sound intensity decreases with the square of the distance from a point source, and prolonged exposure to high sound levels can damage hearing. The logarithmic response of the ear makes the decibel scale a natural and practical way to quantify sound levels.

## 1.11.4 Sources of Sound: Vibrating Strings and Air Columns

Most sounds in nature and music originate from **vibrating systems** such as stretched strings, membranes, air columns, or solid bodies, which act as sources of longitudinal sound waves in air. The specific vibration patterns of these systems determine the frequencies produced and give rise to the concepts of **resonance**, **natural modes**, and **harmonics**.

### Vibrating Strings

A stretched string fixed at both ends can support **standing waves** only if the displacement at each end is zero, meaning that both ends are **nodes**. This boundary condition restricts the motion of the string to a discrete set of allowed vibration patterns, called the natural modes. For a string of length $L$, the allowed wavelengths are
$$
\lambda_n = \frac{2L}{n}, \quad n = 1,2,3,\dots
$$
and the corresponding frequencies are
$$
f_n = \frac{v}{\lambda_n} = \frac{n v}{2L}.
$$
These frequencies form a harmonic series, where the lowest frequency $f_1$ is the **fundamental**, and higher frequencies are called **overtones** or **harmonics**.

### Wave Speed on a String

The speed of transverse waves on a stretched string depends on the tension $T$ and the linear mass density $\mu$ according to
$$
v = \sqrt{\frac{T}{\mu}}.
$$
Substituting this expression into the harmonic frequencies shows that
$$
f_n = \frac{n}{2L}\sqrt{\frac{T}{\mu}}.
$$
This relation explains how the pitch of a string can be controlled in practice, since increasing the tension raises the frequency, increasing the mass per unit length lowers the frequency, and shortening the vibrating length increases the frequency.

### Harmonics and Overtones

Each allowed standing-wave pattern on a string corresponds to a distinct harmonic. The fundamental mode has the longest wavelength and lowest frequency, while higher harmonics have shorter wavelengths and higher frequencies that are integer multiples of the fundamental. The presence and relative strengths of these harmonics determine the richness and character of the sound produced by the string.

### Vibrating Air Columns

Air columns in tubes produce sound by supporting **longitudinal standing waves** of air motion. The allowed standing-wave patterns depend on whether the tube ends are open or closed, because these conditions determine whether particle displacement is a node or an antinode at each end.

### Tubes Open at Both Ends

In a tube open at both ends, the air is free to move at each end, so both ends are **displacement antinodes**. The allowed wavelengths are
$$
\lambda_n = \frac{2L}{n}, \quad n = 1,2,3,\dots
$$
and the corresponding frequencies are
$$
f_n = \frac{n v}{2L}.
$$
All harmonics are present in this case, in direct analogy with a string fixed at both ends.

### Tubes Closed at One End

In a tube closed at one end and open at the other, the closed end is a **node** and the open end is an **antinode**. The allowed wavelengths are
$$
\lambda_n = \frac{4L}{n}, \quad n = 1,3,5,\dots
$$
with corresponding frequencies
$$
f_n = \frac{n v}{4L}, \quad n = 1,3,5,\dots
$$
Only **odd harmonics** are allowed because the boundary conditions cannot accommodate even multiples of half-wavelengths.

### End Correction

In real air columns, the displacement antinode at an open end lies slightly outside the physical end of the tube. To account for this effect, an **end correction** is applied to the effective length of the tube. For a tube of radius $r$, the effective length is approximately
$$
L_{\text{eff}} = L + 0.6r
$$
for a tube open at one end, and
$$
L_{\text{eff}} = L + 1.2r
$$
for a tube open at both ends. This correction slightly lowers the resonant frequencies compared to ideal predictions.

### Resonance and Driving Frequency

When a vibrating source such as a tuning fork or loudspeaker drives a string or air column at one of its natural frequencies, **resonance** occurs and the amplitude of vibration increases significantly. In musical instruments, resonance determines the fundamental pitch, while harmonics contribute to the sound’s timbre. Musicians control resonance by adjusting parameters such as string length, tension, or the effective length of an air column.

### Comparison of Strings and Air Columns

Vibrating strings fixed at both ends and air columns open at both ends share the same harmonic structure, allowing all harmonics. Air columns closed at one end differ fundamentally in that they support only odd harmonics due to their asymmetric boundary conditions. In all cases, the allowed frequencies are determined by the wave speed in the medium and the geometric constraints imposed by the boundaries.

### Conceptual Notes

Vibrating strings and air columns produce sound through standing waves determined by boundary conditions. Strings with nodes at both ends support all harmonics, while tubes closed at one end support only odd harmonics. The wave speed on a string depends on the square root of the ratio of tension to mass density, while air-column resonance depends on sound speed, effective length, and end conditions. Resonance leads to large vibration amplitudes when the driving frequency matches a natural frequency and is fundamental to sound production in musical instruments.

## 1.11.5 Quality of Sound, and Noise; Superposition

Although two sounds may have the same pitch and loudness, they can still be perceived as different, such as the same musical note played on a violin and a flute, and this difference is described as the **quality** or **timbre** of sound. Timbre is determined by the waveform of the sound and by the harmonic components present in the vibration, while irregular and nonperiodic sounds are classified as **noise**, and both musical sound and noise can be understood using the **principle of superposition**.

### Quality (Timbre) of Sound

The quality of a sound is the property that allows us to distinguish between sounds that have the same frequency and loudness. It depends on the number of harmonics present, their relative amplitudes, and their phase relationships with respect to the fundamental frequency. Every vibrating source produces a complex waveform that can be represented as a combination of simple sinusoidal waves, and this specific combination defines the characteristic timbre of the sound.

### Fourier’s Theorem and Harmonic Content

According to Fourier’s theorem, any periodic sound wave can be expressed as a sum of a fundamental sinusoidal oscillation and a series of higher-frequency components called harmonics or overtones, which can be written as
$$
y(t) = A_1 \sin(\omega t) + A_2 \sin(2\omega t + \phi_2) + A_3 \sin(3\omega t + \phi_3) + \dots
$$
where the first term represents the fundamental frequency and the higher-order terms represent harmonics at integer multiples of the fundamental frequency. The amplitudes and phases of these components determine the overall waveform shape and therefore the perceived sound quality.

### Harmonic Structure and Instrumental Timbre

Different musical instruments emphasize different harmonics, which explains why they sound distinct even when playing the same note. Instruments such as flutes produce waveforms close to a pure sine wave dominated by the fundamental frequency, giving a soft and mellow sound, while instruments like violins produce many strong harmonics, resulting in a richer and brighter tone. Instruments that emphasize mainly odd harmonics, such as closed organ pipes or clarinets, produce waveforms similar to square waves and are perceived as hollow or reedy.

### Musical Notes and Frequency Ratios

Musical harmony is closely related to simple ratios of frequencies between notes. Intervals such as octaves, fifths, and thirds correspond to simple integer ratios, which produce consonant and pleasing sounds, while more complex ratios lead to dissonance. These relationships arise naturally from the harmonic structure of vibrating systems.

### Noise

Noise is produced by irregular, nonperiodic vibrations that do not have a definite frequency or harmonic structure. Unlike musical tones, noise contains a broad range of unrelated frequencies with random phase relationships. Common examples include traffic sounds, rustling leaves, and radio static. Noise is typically described statistically in terms of its average intensity or frequency spectrum rather than by a single pitch.

### Superposition of Sound Waves

When multiple sound waves occupy the same region of space, their displacements add according to the principle of superposition, so that the total displacement is the sum of the individual contributions. Depending on their relative phases and frequencies, superposition can lead to interference, beats, or resonance. Coherent superposition with a fixed phase relationship produces interference effects, while incoherent superposition with random phases leads to steady combined intensities, which is characteristic of noise.

### Waveform Shape and Perceived Tone

The human ear responds to the overall waveform rather than to instantaneous pressure variations. As a result, the shape of the waveform determines whether a sound is perceived as smooth, harsh, bright, or dull. A pure sinusoidal waveform corresponds to a simple tone, whereas complex waveforms containing many harmonics produce richer and more textured sounds.

### Musical Scales and Frequency Spacing

In the equal-tempered musical scale used in Western music, an octave corresponding to a frequency ratio of 2:1 is divided into twelve equal logarithmic steps. Each successive note has a frequency given by
$$
f_n = f_0 2^{n/12},
$$
which allows music to be played consistently in different keys, even though the resulting intervals are only approximate representations of simple harmonic ratios.

### Sound Envelopes and Amplitude Modulation

Real musical sounds vary in amplitude over time and are characterized by an envelope consisting of attack, decay, sustain, and release phases. These temporal variations, together with harmonic content, strongly influence the perceived character of a sound and help distinguish different instruments even when they play the same note.

### Musical Tone versus Noise

Musical tones are produced by regular, periodic waveforms with definite frequencies and harmonically related overtones, whereas noise arises from irregular, nonperiodic waveforms with a broad and unstructured frequency spectrum. The ear generally perceives musical tones as organized and pleasant, while noise is perceived as harsh or unstructured.

### Conceptual Notes

Timbre distinguishes sounds of equal pitch and loudness and is determined by harmonic content and waveform shape. Fourier analysis provides the mathematical framework for decomposing complex sounds into simple sinusoidal components. Noise differs fundamentally from musical sound because it lacks periodicity and harmonic structure. The superposition principle underlies the formation of all complex sounds, and time-dependent amplitude envelopes play a crucial role in musical expressiveness.

In the next section, we study the **interference of sound waves** in more detail and examine how slightly different frequencies give rise to **beats**.


## 1.11.6 Interference of Sound Waves; Beats

When two or more sound waves occupy the same region in space, they combine according to the **principle of superposition**, and this combination can produce regions of increased or decreased amplitude, leading to **interference** and **beats**. Both effects arise from wave superposition, but interference involves waves of the same frequency, whereas beats occur when the frequencies differ slightly.

### Interference of Sound Waves

When two coherent sound waves with the same frequency $f$ and amplitude $A$ overlap, their displacements add algebraically. For
$$
y_1 = A \sin(\omega t), \qquad y_2 = A \sin(\omega t + \phi),
$$
where $\phi$ is the phase difference, their superposition is
$$
y = 2A \cos\!\left(\frac{\phi}{2}\right)\sin\!\left(\omega t + \frac{\phi}{2}\right).
$$
The resulting amplitude is therefore
$$
A_R = 2A \cos\!\left(\frac{\phi}{2}\right),
$$
showing that the amplitude depends entirely on the phase difference between the waves.

### Constructive and Destructive Interference

If the phase difference is $\phi = 0, 2\pi, 4\pi, \dots$, the waves interfere constructively and the amplitude reaches its maximum value $A_R = 2A$. If $\phi = \pi, 3\pi, 5\pi, \dots$, destructive interference occurs and the resultant amplitude is zero. When the waves travel different path lengths, the phase difference is related to the path difference $\Delta x$ by
$$
\phi = \frac{2\pi}{\lambda}\Delta x,
$$
so constructive interference occurs for $\Delta x = n\lambda$ and destructive interference for $\Delta x = (n+\tfrac12)\lambda$.

### Beats: Interference of Slightly Different Frequencies

When two sound waves of nearly equal frequencies $f_1$ and $f_2$ interfere, the combined sound alternates between loud and soft, producing **beats**. For
$$
y_1 = A \sin(\omega_1 t), \qquad y_2 = A \sin(\omega_2 t),
$$
with $\omega_i = 2\pi f_i$ and $f_1 \approx f_2$, their sum can be written as
$$
y = 2A \cos\!\left[\frac{(\omega_1-\omega_2)t}{2}\right]\sin\!\left[\frac{(\omega_1+\omega_2)t}{2}\right].
$$
The sine term represents a tone at the average frequency $(f_1+f_2)/2$, while the cosine term modulates the amplitude at the **beat frequency**
$$
f_{\text{beat}} = |f_1 - f_2|.
$$

### Physical Interpretation of Beats

During beats, the sound intensity varies periodically between maxima, when the waves interfere constructively, and minima, when they interfere destructively. The ear perceives these periodic intensity variations as fluctuations in loudness. If the frequency difference is 1 Hz, one beat per second is heard, while larger differences eventually produce a rough or dissonant sound rather than distinct beats.

### Intensity Variation During Beats

Because sound intensity is proportional to the square of the amplitude, the instantaneous intensity during beats varies as
$$
I(t) \propto [2A \cos(\pi \Delta f\, t)]^2 = 4A^2 \cos^2(\pi \Delta f\, t),
$$
showing that the intensity oscillates between zero and a maximum value at the beat frequency.

### Beats as a Measurement Tool

Beats provide a sensitive method for comparing frequencies. When a sound source of unknown frequency is compared with a known reference, beats are heard if the frequencies differ. By adjusting the unknown source until the beats disappear, the two frequencies are made equal, allowing precise tuning and frequency determination.

### Interference and Beats Compared

Interference occurs when waves of the same frequency overlap and produces spatial regions of loudness and silence determined by path differences. Beats occur when frequencies differ slightly and produce temporal variations in loudness determined by the frequency difference. Both effects are direct consequences of wave superposition and are widely used in acoustics and music.

### Conceptual Notes

Interference and beats both arise from the superposition principle applied to sound waves. Constructive interference produces loud regions, while destructive interference produces quiet regions. Beats occur only when frequencies differ slightly, with a beat frequency equal to the absolute difference of the component frequencies. Beats are essential for musical tuning and precise frequency measurement. Interference can manifest either in space, as with multiple sound sources, or in time, as with beats.

In the next section, we examine the **Doppler effect**, which describes how the observed frequency of sound changes when the source or the observer is in motion.


## 1.11.7 Doppler Effect

The **Doppler effect** is the change in the **observed frequency or wavelength** of a wave caused by **relative motion** between the source and the observer, as commonly experienced in the changing pitch of a passing siren where the sound is higher during approach and lower during recession. This effect occurs for all types of waves, including sound and electromagnetic waves, and here it is introduced for sound.

### Principle of the Doppler Effect

The observed change in frequency arises because relative motion changes the rate at which successive wavefronts reach the observer. When the source approaches the observer, wavefronts are compressed and the observed frequency increases, whereas when the source recedes, wavefronts are stretched and the observed frequency decreases. The emitted frequency of the source remains constant, but the frequency perceived by the observer changes.

### Doppler Effect for Sound Waves

Let $v$ denote the speed of sound in the medium, $v_O$ the speed of the observer taken as positive when moving toward the source, $v_S$ the speed of the source taken as positive when moving away from the observer, $f_S$ the frequency emitted by the source, and $f_O$ the frequency observed. The general Doppler formula for sound is
$$
f_O = f_S \frac{v + v_O}{v - v_S}.
$$

### Special Cases

If the source moves while the observer is stationary, so that $v_O = 0$, the observed frequency is
$$
f_O = f_S \frac{v}{v - v_S},
$$
which increases when the source approaches and decreases when it recedes. If the observer moves while the source is stationary, so that $v_S = 0$, the observed frequency becomes
$$
f_O = f_S \frac{v + v_O}{v},
$$
which again increases when the observer moves toward the source and decreases when moving away. If both the source and observer move, the general expression applies directly.

### Wavelength Interpretation

From the wavelength perspective, the motion of the source modifies the wavelength emitted into the medium according to
$$
\lambda' = \frac{v - v_S}{f_S}.
$$
If the observer moves relative to the medium, the frequency detected is
$$
f_O = \frac{v \pm v_O}{\lambda'},
$$
which leads to the same Doppler result as the frequency-based formulation.

### Approaching and Receding Source Example

For a car horn emitting $f_S = 500~\text{Hz}$ and moving toward a stationary observer at $v_S = 30~\text{m/s}$ with sound speed $v = 343~\text{m/s}$, the observed frequency is
$$
f_O = 500 \frac{343}{343 - 30} = 549~\text{Hz},
$$
corresponding to a higher perceived pitch. If the same source moves away, the observed frequency becomes
$$
f_O = 500 \frac{343}{343 + 30} = 460~\text{Hz},
$$
corresponding to a lower perceived pitch.

### Moving Source and Moving Observer

When both the source and observer are in motion, their effects combine. For example, if a source emits $700~\text{Hz}$ while moving at $30~\text{m/s}$ toward an observer who runs toward the source at $10~\text{m/s}$, the observed frequency is
$$
f_O = 700 \frac{343 + 10}{343 - 30} = 789~\text{Hz},
$$
showing a significant upward shift.

### Doppler Effect for Reflected Sound

When sound reflects from a moving object, such as in radar or ultrasound, the Doppler shift occurs twice, once during reception by the moving object and again during re-emission. For a target moving at speed $u$ much smaller than $v$, the reflected frequency is approximately
$$
f' \approx f_S \left(1 + \frac{2u}{v}\right),
$$
which is known as the **double Doppler shift** and is widely used in medical and technical applications.

### Small Velocity Approximation

If the source and observer speeds are much smaller than the speed of sound, the fractional frequency shift can be approximated by
$$
\frac{\Delta f}{f_S} \approx \frac{v_O - v_S}{v}.
$$
This linear approximation shows that the Doppler shift is directly proportional to relative velocity.

### Doppler Effect for Light

For light waves, which do not require a medium, the Doppler effect must be treated relativistically. The observed frequency is given by
$$
f_O = f_S \sqrt{\frac{1 + \beta}{1 - \beta}}, \quad \beta = \frac{v}{c},
$$
leading to a **blue shift** when the source approaches and a **red shift** when it recedes. This effect is fundamental in astronomy for determining stellar and galactic motion.

### Conceptual Notes

The Doppler effect results from relative motion between a source and an observer and leads to an increase in frequency when they approach each other and a decrease when they separate. The wave speed in the medium remains constant while wavelength and observed frequency change. For small velocities, the frequency shift is proportional to relative speed divided by wave speed. The Doppler effect applies to sound, light, and other waves and underlies technologies such as radar, sonar, medical ultrasound, and astronomical spectroscopy.

In the next section, we examine **shock waves and the sonic boom**, which arise when an object moves faster than the speed of sound in a medium.


## 1.11.8 Shock Waves and the Sonic Boom

When a source moves **faster than the speed of sound** in a medium, the emitted wavefronts overlap and form a region of **constructive interference** known as a **shock wave**. For an observer, this sudden compression is perceived as a **sonic boom**, which is an intense and brief sound pulse associated with supersonic motion.

### Sound Waves from a Moving Source

If a sound source moves at a speed **less than** the speed of sound $v_S < v$, each successive wavefront is emitted from a slightly shifted position, causing wavefronts to crowd in the forward direction and producing a Doppler shift. As the source speed approaches the sound speed, the spacing between wavefronts in front of the source decreases dramatically and the wave geometry changes qualitatively.

### Source Moving at the Speed of Sound

When the source speed equals the speed of sound $v_S = v$, each new wavefront is emitted from a point on the same spherical surface as the previous one, so the wavefronts pile up at the front of the source. In theory this leads to infinite compression, while in practice it produces extremely large pressure gradients. This condition defines **Mach 1**, the threshold of supersonic motion.

### Supersonic Motion and Shock Waves

When the source moves faster than sound $v_S > v$, each wavefront is emitted ahead of the earlier ones, and the overlapping wavefronts form a conical envelope. This envelope is a **shock wave**, characterized by abrupt increases in pressure, density, and temperature. The shock wave propagates outward as a sharply defined front rather than as a smooth oscillation.

### Mach Number and Mach Angle

The **Mach number** is defined as
$$
M = \frac{v_S}{v}.
$$
The half-angle $\theta$ of the resulting shock cone, known as the **Mach angle**, satisfies
$$
\sin \theta = \frac{v}{v_S} = \frac{1}{M}.
$$
As the Mach number increases, the cone becomes narrower, concentrating the shock wave closer to the direction of motion.

### Mach Cone Example

For a jet flying at $v_S = 510~\text{m/s}$ where the speed of sound is $v = 340~\text{m/s}$, the Mach number is
$$
M = \frac{510}{340} = 1.5,
$$
and the Mach angle follows from
$$
\sin \theta = \frac{1}{1.5} = 0.667,
$$
giving $\theta \approx 42^\circ$. The aircraft therefore generates a Mach cone with a half-angle of about $42^\circ$.

### The Sonic Boom

As the shock wave propagates outward and intersects the ground, an observer experiences a sudden pressure change perceived as a **sonic boom**. The boom is not produced only at the instant the source exceeds Mach 1 but continuously along the flight path of the supersonic object. Each observer hears the boom when the Mach cone sweeps past their location.

### Pressure Structure of a Sonic Boom

The pressure variation associated with a sonic boom typically has an **N-wave** profile consisting of a sharp pressure rise at the leading shock, followed by a gradual decrease below ambient pressure, and a second shock returning the pressure to normal. Typical overpressures are on the order of 20–100 Pa, which can rattle windows and produce loud noise but usually do not cause structural damage.

### Shock Waves in Fluids

Shock waves occur in any compressible medium, not only in air. They are marked by irreversible changes in pressure, density, and temperature and cannot be described by linear wave theory. Examples include explosions, thunder from lightning, detonations, and shock fronts generated by supersonic projectiles or spacecraft.

### Supersonic and Hypersonic Flight

Near Mach 1, aerodynamic drag rises sharply, a phenomenon historically called the **sound barrier**. Supersonic flight corresponds to $1 < M < 5$, while **hypersonic** flight occurs for $M > 5$. At hypersonic speeds, intense heating and even ionization of air can occur around the vehicle, leading to plasma formation.

### Energy Dissipation in Shock Waves

Shock waves are inherently **dissipative**, meaning that mechanical energy is converted into internal energy of the medium. This leads to heating and entropy increase, distinguishing shock waves from ordinary sound waves, which are linear, reversible, and transport energy without permanent changes to the medium.

### Conceptual Notes

Shock waves form when a source moves faster than the wave speed in the medium, producing a Mach cone defined by $\sin \theta = v / v_S$. The **sonic boom** is the audible manifestation of the shock front reaching an observer. The **Mach number** quantifies the degree of supersonic motion. Unlike ordinary sound waves, shock waves involve sudden, nonlinear changes in pressure, density, and temperature and play an important role in aerodynamics, explosions, and high-speed flight physics.

In the next section, we examine how sound waves are exploited in **applications** such as sonar, ultrasound, and medical imaging, where reflection and Doppler effects are used for measurement and diagnosis.


## 1.11.9 Applications: Sonar, Ultrasound, and Medical Imaging

The study of sound waves — particularly their reflection, refraction, and Doppler behavior — has led to powerful practical applications. Among the most important are **sonar** for underwater detection, **ultrasound** for industrial inspection, and **medical ultrasound imaging** for noninvasive diagnostics.

### Sonar (Sound Navigation and Ranging)

**Sonar** systems locate underwater objects by emitting sound pulses and detecting their **echoes** after reflection. Because the speed of sound in water is finite (about 1500 m/s), the time delay between emission and reception can be used to determine distance.

If a pulse returns after a time interval $\Delta t$, the distance to the reflecting object is
$$
d = \frac{1}{2} v \Delta t,
$$
where $v$ is the speed of sound in water. The factor of one-half accounts for the round trip of the sound pulse.

#### Example: Measuring Ocean Depth

A sonar pulse returns from the ocean floor after $1.20~\text{s}$. With $v = 1500~\text{m/s}$,
$$
d = \frac{1}{2}(1500)(1.20) = 900~\text{m}.
$$
The ocean depth is therefore **900 m**.

### Types of Sonar

Two main sonar modes are used in practice:

| Type | Description | Typical Use |
|------|-------------|-------------|
| Active sonar | Emits sound pulses and listens for echoes | Mapping, fish finding |
| Passive sonar | Detects sound produced by other sources | Submarine detection, marine biology |

In addition to distance, sonar systems can determine **relative speed** using the Doppler effect.

### Doppler Sonar and Velocity Measurement

If the reflecting object moves with speed $u$, the reflected frequency differs from the emitted one due to a **double Doppler shift**. For $u \ll v$,
$$
f' \approx f_S \left(1 + \frac{2u}{v}\right).
$$
By measuring this frequency shift, sonar can determine the velocity of moving objects such as submarines or schools of fish.

### Ultrasound

**Ultrasound** refers to sound waves with frequencies above $20~\text{kHz}$, beyond the range of human hearing. Practical applications typically use frequencies from **1 to 15 MHz**.

High frequencies correspond to short wavelengths, allowing ultrasound to resolve small structures and interfaces. This makes ultrasound valuable for both **industrial testing** and **medical imaging**.

### Generation and Detection of Ultrasound

Ultrasound is generated using **piezoelectric crystals**. When an alternating voltage is applied, the crystal vibrates mechanically at the same frequency and emits ultrasonic waves. When driven at resonance, the emission is especially strong.

The same crystal can act as a receiver: incoming ultrasound waves deform the crystal and generate an electrical signal. This dual role allows compact transmitter–receiver systems.

### Industrial Applications of Ultrasound

Ultrasound is widely used in industry for tasks requiring precision and safety:

- **Non-destructive testing:** detecting internal cracks or voids in materials,
- **Ultrasonic cleaning:** removing contaminants via cavitation,
- **Welding and cutting:** joining plastics or biological tissues,
- **Level and flow measurement:** monitoring fluids without direct contact.

### Medical Uses of Ultrasound

In medicine, ultrasound is especially valuable because it is **nonionizing** and **noninvasive**. Imaging is based on partial reflections at boundaries between tissues with different acoustic properties.

Typical medical applications include:

| Application | Frequency Range | Purpose |
|-------------|-----------------|---------|
| Obstetric imaging | 3–5 MHz | Fetal visualization |
| Echocardiography | 2–7 MHz | Heart motion and valves |
| Abdominal imaging | 2–10 MHz | Internal organs |
| Doppler vascular imaging | 5–10 MHz | Blood flow measurement |

### Example: Ultrasound Distance Measurement

An ultrasound pulse travels through soft tissue at $v = 1540~\text{m/s}$. If the echo returns after $\Delta t = 130~\mu\text{s}$,
$$
d = \frac{1}{2}(1540)(130 \times 10^{-6}) = 0.10~\text{m}.
$$
The reflecting boundary lies **10 cm** below the surface.

### Doppler Ultrasound and Blood Flow

Doppler ultrasound measures blood velocity by detecting the frequency shift caused by moving blood cells. The approximate relation is
$$
\Delta f = 2 f_S \frac{u}{v} \cos \theta,
$$
where $\theta$ is the angle between the ultrasound beam and the direction of flow.

This technique allows clinicians to:
- Measure blood flow speed,
- Identify arterial narrowing or blockage,
- Assess heart valve performance.

#### Example: Blood Velocity from Doppler Shift

For $f_S = 5.0~\text{MHz}$, $\Delta f = 3.3~\text{kHz}$, $v = 1540~\text{m/s}$, and $\theta = 0°$,
$$
u = \frac{\Delta f \, v}{2 f_S} = 0.51~\text{m/s}.
$$
The blood flow speed is approximately **0.5 m/s**.

### Resolution and Frequency Tradeoff

In ultrasound imaging, resolution is limited roughly by one wavelength:
$$
\Delta x \approx \frac{v}{f}.
$$
Higher frequencies improve spatial resolution but reduce penetration depth. Lower frequencies penetrate deeper but with reduced detail. Medical imaging balances these effects depending on the target tissue.

### Safety and Advantages of Ultrasound

Diagnostic ultrasound has several major advantages:
- No ionizing radiation,
- Real-time imaging,
- Portable and cost-effective equipment,
- Safe for routine clinical use.

At very high intensities, ultrasound can cause heating or cavitation and is used therapeutically for procedures such as tissue ablation or kidney stone fragmentation.

### Conceptual Notes

- Sonar determines distance and motion using reflected sound waves in water.
- Ultrasound employs high-frequency sound for fine spatial resolution.
- Medical ultrasound relies on reflection at tissue boundaries.
- Doppler techniques extract velocity information from frequency shifts.
- Frequency determines the tradeoff between resolution and penetration.
- Ultrasound is a versatile and widely used diagnostic and industrial tool.
