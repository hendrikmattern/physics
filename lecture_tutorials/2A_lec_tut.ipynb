## From vectors to vector fields (2D example)

**Vector**: Magnitude and direction, visualized as arrows

$$
\vec{v} = v_x \hat{i} + v_y \hat{j}
\qquad \text{or} \qquad
\vec{v} =
\begin{bmatrix}
v_x \\[4pt]
v_y
\end{bmatrix}
$$

**Vector field**: Assigns a vector to every point in space

$$
\vec{F}(x, y) = -y \,\hat{i} + x \,\hat{j}
\qquad \text{or} \qquad
\vec{F}(x, y) =
\begin{bmatrix}
- y \\[4pt]
x
\end{bmatrix}

$$

$\Rightarrow$ **At each point $(x, y)$, the field assigns a vector whose components depend on position.**

Example:

At $(1, 0): \vec{F} = (0, 1)$  
At $(0, 1): \vec{F} = (-1, 0)$  



## Vector fields: How to visualize

Two complementary representations help us understand a vector field:

* **Arrows (local view):** vectors at *selected points* show magnitude and direction  
* **Field lines (global view):** *continuous* curves reveal the overall structure and flow

## Line integrals of scalar fields

A scalar line integral accumulates a scalar quantity along a path: 
$$\int_{\mathcal{C}} f \, ds$$

* Idea:
  * Split path $\mathcal{C}$ into small elements $ds$  
  * $ds$ measures arc length along the curve  
  * Sum values $f \, ds$ along the path  
* No direction dependence, only magnitude matters, but $f$ can vary along the path and may be positive or negative
* Result is a single scalar quantity representing the total amout along the path
* Scalar line integrals generalize $\int f(x)\,dx$ to curves in space

## Line integrals of vector fields

A vector line integral measures how a field acts along a path:
$$
\int_{\mathcal{C}} \vec{F} \cdot d\vec{l}
$$

* Extends scalar line integrals by considering direction of path and vector field  
* Idea:
  * Split path $\mathcal{C}$ into small segments $d\vec{l}$  
  * $d\vec{l}$ points along the path  
  * Sum contributions $\vec{F} \cdot d\vec{l}$ along the path  
* Direction matters through the dot product, only the component along the path contributes  
* Result is a single scalar measuring the total effect along the path

## Closed line integrals & conservative fields

* **Closed line integral** follows a path that starts and ends at the same point:
$$
\oint \vec{F} \cdot d\vec{l}
$$
* For **conservative fields**:
$$
\Rightarrow \oint \vec{F} \cdot d\vec{l} = 0
$$  
* For a closed loop, start and end coincide  
* **Intuition:** moving “forward” and coming back cancels exactly $\Rightarrow$ no net effect remains
* $\Rightarrow$ **Path independence** for conservative fields
* If a closed integral is **non-zero** $\Rightarrow$ the field has circulation (not conservative)  
  

## Flux & surface integrals: How much field passes through a surface

Flux $\Phi$ measures how much of a vector field $\vec{F}$ flows through a surface $\mathcal{S}$:
$$
\Phi = \int_{\mathcal{S}} \vec{F} \cdot d\vec{A}
$$
* Flux is the surface integral of a vector field  
* **Idea:**
  * Split surface $\mathcal{S}$ into small patches $d\vec{A} = \hat{n}\,dA$  (normal $\hat{n}$)
  * Dot product $\Rightarrow$ only component through surface contributes  
  * Sum contributions $\vec{F} \cdot d\vec{A}$ over the surface  
* Field aligned with normal $\rightarrow$ maximum  
* Field parallel to surface $\rightarrow$ zero  
* Result is a single scalar: total flow through the surface

## Gradient: From scalar field to direction of change

The **gradient** $\nabla V$ points in the direction of strongest increase of $V(x,y,z)$, thus converts a scalar field into a vector field:
$$
\nabla V
$$

$$
\nabla V =
\begin{bmatrix}
\frac{\partial V}{\partial x} \\[6pt]
\frac{\partial V}{\partial y} \\[6pt]
\frac{\partial V}{\partial z}
\end{bmatrix}
\qquad \text{or} \qquad
\nabla V = 
\frac{\partial V}{\partial x}\,\hat{i} +
\frac{\partial V}{\partial y}\,\hat{j} +
\frac{\partial V}{\partial z}\,\hat{k}
$$

* **Idea:**
  * Measure how $V$ changes in each spatial direction, i.e. take the derivative 
  * Combine spatial derivatives into a vector $\rightarrow$ **for each point, we know the direction of strongest increase**
* Result is a vector field describing change in space  
* $\nabla V$ is perpendicular to surfaces of constant $V$  
* Magnitude: $|\nabla V|$ gives strength of change

## Example: Gradient of a simple scalar field

Given a scalar field:
$$
V(x,y) = x^2 + y^2
$$

Compute the gradient:
$$
\nabla V =
\begin{bmatrix}
2x \\
2y
\end{bmatrix}
$$

* Gradient points radially outward  
* Magnitude increases with distance from origin  
* Direction = steepest increase of $V$  

## From scalar to vector field derivatives

We can take derivatives of scalar fields:
$$
\nabla V(x,y,z) \quad \text{(gradient)}
$$

What about vector fields $\vec{F}(x,y,z)$?

* A vector field has multiple components  
  $\Rightarrow$ derivatives can be combined in different ways  
* Apply $\nabla$ to a vector field:
  * Dot product $\Rightarrow$ divergence $\nabla \cdot \vec{F}$  
  * Cross product $\Rightarrow$ curl $\nabla \times \vec{F}$  
* These operations extract different physical information  
  * Divergence: measures sources and sinks  
  * Curl: measures rotation of the field  

## Divergence: Sources and sinks of a field

Divergence measures how a vector field spreads out or converges:
$$
\nabla \cdot \vec{F}
$$
$\Rightarrow$ **Differentiate each component along its own direction and sum the results**
* **Conceptual idea:**
  * Consider a small volume around a point  
  * Compare inflow and outflow through the surface  
  * Assign net outflow to the point  
  * Positive $\rightarrow$ source (flow outward)  
  * Negative $\rightarrow$ sink (flow inward)  
  * Zero $\rightarrow$ no net source or sink  
* **Result is a scalar field describing local source strength**

*<sub>from [wikipedia](https://en.wikipedia.org/wiki/File:Divergence_(captions).svg), [Attribution-Share Alike 4.0 International](//creativecommons.org/licenses/by-sa/4.0/deed.en)</sub>*

## Example: Divergence of a vector field

Given a vector field:
$$
\vec{F}(x,y) =
\begin{bmatrix}
x^2 \\
2y + 1
\end{bmatrix}
$$

Compute the divergence:
$$
\nabla \cdot \vec{F}
=
\frac{\partial F_x}{\partial x}
+
\frac{\partial F_y}{\partial y}
=
\frac{\partial x^2}{\partial x}
+
\frac{\partial (2y+1)}{\partial y}
= 2x + 2
$$

* Divergence depends on position
* Positive for $x > -1$ $\rightarrow$ source  
* Negative for $x < -1$ $\rightarrow$ sink  
* Results is a **scalar field** (single number which can depend on position but does not have direction/components)

## Curl: Local rotation of a field

Curl measures how a vector field $\vec{F}(x,y,z)$  rotates or circulates:
$$
\nabla \times \vec{F} =
\begin{vmatrix}
\hat{i} & \hat{j} & \hat{k} \\
\partial_x & \partial_y & \partial_z \\
F_x & F_y & F_z
\end{vmatrix}
=
\begin{bmatrix}
\frac{\partial F_z}{\partial y} - \frac{\partial F_y}{\partial z} \\
\frac{\partial F_x}{\partial z} - \frac{\partial F_z}{\partial x} \\
\frac{\partial F_y}{\partial x} - \frac{\partial F_x}{\partial y}
\end{bmatrix}
$$
$\Rightarrow$ **Combine derivatives of components across directions via a cross product**

* **Conceptual idea:**
  * Consider a tiny paddle wheel at a point  
  * Check whether the field causes rotation  
  * Direction $\rightarrow$ axis of rotation (right-hand rule)  
  * Magnitude $\rightarrow$ strength of rotation  
  * Zero $\rightarrow$ no local rotation  

* **Result is a vector field describing local circulation**

## Example 1: Curl of a vector field

Given vector field from divergence example:
$$
\vec{F}(x,y) =
\begin{bmatrix}
x^2 \\
2y + 1
\end{bmatrix}

\quad \Rightarrow \quad 

\vec{F}(x,y,z) =
\begin{bmatrix}
x^2 \\
2y + 1 \\
0
\end{bmatrix}

$$

*Note: The curl points perpendicular to the plane of local rotation, not necessarily perpendicular to the vector field itself*

Curl is defined as a cross product with $\nabla$:
$$
\nabla \times \vec{F}
=
\begin{vmatrix}
\hat{i} & \hat{j} & \hat{k} \\
\partial_x & \partial_y & \partial_z \\
x^2 & 2y+1 & 0
\end{vmatrix}
=
\begin{bmatrix}
\frac{\partial 0}{\partial y} - \frac{\partial (2y+1)}{\partial z} \\
\frac{\partial x^2}{\partial z} - \frac{\partial 0}{\partial x} \\
\frac{\partial (2y+1)}{\partial x} - \frac{\partial x^2}{\partial y}
\end{bmatrix}
$$

$$
=
\hat{k}\left(
\frac{\partial (2y+1)}{\partial x}
-
\frac{\partial x^2}{\partial y}
\right)
=
\hat{k}(0 - 0)
= 0
$$

* Curl is zero $\rightarrow$ no local rotation  
* Divergence $\neq 0$ $\rightarrow$ sources/sinks present  
* $\Rightarrow$ Spreading $\neq$ rotating  


## Example 2: Curl of a vector field

Given vector field:
$$
\vec{F}(x,y,z) =
\begin{bmatrix}
y \\
-x \\
0
\end{bmatrix}

$$

Curl is defined as a cross product with $\nabla$:
$$
\nabla \times \vec{F}
=
\begin{vmatrix}
\hat{i} & \hat{j} & \hat{k} \\
\partial_x & \partial_y & \partial_z \\
y & -x & 0
\end{vmatrix}
=
\begin{bmatrix}
\frac{\partial 0}{\partial y} - \frac{\partial (-x)}{\partial z} \\
\frac{\partial y}{\partial z} - \frac{\partial 0}{\partial x} \\
\frac{\partial (-x)}{\partial x} - \frac{\partial y}{\partial y}
\end{bmatrix}
=
\hat{k}\left(
\frac{\partial (-x)}{\partial x}
-
\frac{\partial y}{\partial y}
\right)
=
\hat{k}(-1 - 1)
= -2\hat{k}
$$

* Result is a **vector field** (here along $-\hat{k}$)  
* Curl is constant and non-zero  
* Field rotates around the origin (clockwise) & field forms closed loops

## Example 3: Curl of a vector field

Given vector field:
$$
\vec{F}(x,y,z) =
\begin{bmatrix}
0 \\
- x^2 \\
0
\end{bmatrix}

$$

Curl is defined as a cross product with $\nabla$:
$$
\nabla \times \vec{F}
=
\begin{vmatrix}
\hat{i} & \hat{j} & \hat{k} \\
\partial_x & \partial_y & \partial_z \\
0 & -x^2 & 0
\end{vmatrix}
=
\begin{bmatrix}
\frac{\partial 0}{\partial y} - \frac{\partial (-x^2)}{\partial z} \\
\frac{\partial 0}{\partial z} - \frac{\partial 0}{\partial x} \\
\frac{\partial (-x^2)}{\partial x} - \frac{\partial 0}{\partial y}
\end{bmatrix}
=
\hat{k}\left(
\frac{\partial (-x^2)}{\partial x}
-
\frac{\partial 0}{\partial y}
\right)
=
\hat{k}(-2x - 0)
= -2x\,\hat{k}
$$

* Curl depends on position  
* Zero at $x = 0$, increases with $|x|$  
* Direction changes sign with $x$  
* Shear flow: velocity varies with position  
  $\Rightarrow$ neighboring layers move differently and induce local rotation  
* Shear flow: Flow speed varies with position $\Rightarrow$ neighboring layers slide past each other $\Rightarrow$ a small paddle wheel would start rotating

## Versatility of the Del / nabla operator

$$
\nabla =
\begin{bmatrix}
\frac{\partial}{\partial x} \\[6pt]
\frac{\partial}{\partial y} \\[6pt]
\frac{\partial}{\partial z}
\end{bmatrix}
\qquad \text{or} \qquad
\nabla =
\hat{i}\,\frac{\partial}{\partial x} +
\hat{j}\,\frac{\partial}{\partial y} +
\hat{k}\,\frac{\partial}{\partial z}
$$

**Gradient:**
$$
\operatorname{grad} f = \nabla f
$$

**Divergence:**
$$
\operatorname{div} \mathbf{v} = \nabla \cdot \mathbf{v}
$$

**Curl:**
$$
\operatorname{curl} \mathbf{v} = \nabla \times \mathbf{v}
$$

## Flux and sources (Gauss' theorem)

The total flux through a closed surface $\mathcal{S}$ equals the sum of sources inside the surface$:
$$
\oint_{\mathcal{S}} \vec{F} \cdot d\vec{A}
=
\int_{\mathcal{V}} (\nabla \cdot \vec{F}) \, dV
$$

* Left: total flux through the surface $\mathcal{S}$  
* Right: divergence integrated over the volume $\mathcal{V}$  
* **Conceptual idea:**
  * Divergence measures local sources (per volume)  
  * Integrating over a volume gives total source strength  
  * Equals total flux through the boundary surface  
  * $\Rightarrow$ Flux = accumulated divergence  

## Circulation and rotation (Stokes' theorem)

The circulation along a closed path equals the total rotation inside:
$$
\oint_{\partial \mathcal{S}} \vec{F} \cdot d\vec{r}
=
\int_{\mathcal{S}} (\nabla \times \vec{F}) \cdot d\vec{A}
$$

* Left: circulation along the boundary $\partial \mathcal{S}$
* Right: curl integrated over the entire surface $\mathcal{S}$
* **Conceptual idea:**
  * Curl measures local rotation (circulation density)  
  * Integrating curl over a surface gives total circulation  
  * local (curl) → integrate → global (circulation)  
  * $\Rightarrow$ Circulation = accumulated curl  

## Why is the curl of a conservative field zero?

* In conservative fields, any closed path has zero circulation:
$$
\oint_{\partial \mathcal{S}} \vec{F} \cdot d\vec{l} = 0
$$

* **Using Stokes' theorem:**
$$
\oint_{\partial \mathcal{S}} \vec{F} \cdot d\vec{l}
=
\int_{\mathcal{S}} (\nabla \times \vec{F}) \cdot d\vec{A}
$$

$$
\Rightarrow
\int_{\mathcal{S}} (\nabla \times \vec{F}) \cdot d\vec{A} = 0
$$

* This must hold for any surface $\mathcal{S}$  
  $\Rightarrow$ the integrand must vanish  

$$
\nabla \times \vec{F} = 0
$$

* **Result: conservative fields are irrotational (no local rotation)**  

## Why is the curl of a gradient zero?

$$
\nabla \times (\nabla V)
=
\begin{bmatrix}
\frac{\partial}{\partial y}\left(\frac{\partial V}{\partial z}\right)
-
\frac{\partial}{\partial z}\left(\frac{\partial V}{\partial y}\right) \\[6pt]
\frac{\partial}{\partial z}\left(\frac{\partial V}{\partial x}\right)
-
\frac{\partial}{\partial x}\left(\frac{\partial V}{\partial z}\right) \\[6pt]
\frac{\partial}{\partial x}\left(\frac{\partial V}{\partial y}\right)
-
\frac{\partial}{\partial y}\left(\frac{\partial V}{\partial x}\right)
\end{bmatrix}
=
\begin{bmatrix}
\underbrace{\frac{\partial^2 V}{\partial y \partial z}
-
\frac{\partial^2 V}{\partial z \partial y}}_{=0} \\[6pt]
\underbrace{\frac{\partial^2 V}{\partial z \partial x}
-
\frac{\partial^2 V}{\partial x \partial z}}_{=0} \\[6pt]
\underbrace{\frac{\partial^2 V}{\partial x \partial y}
-
\frac{\partial^2 V}{\partial y \partial x}}_{=0}
\end{bmatrix}
=
\vec{0}
$$


* Order of differentiation does not matter, $\partial_i \partial_j V = \partial_j \partial_i V$
* $\Rightarrow$ all terms cancel pairwise  
* $\Rightarrow$ no local rotation  
* **Result: $\nabla \times (\nabla V) = 0$ (gradient fields are irrotational)**

## Why is the divergence of a curl zero?

$$
\nabla \cdot (\nabla \times \vec{F})
=
\frac{\partial}{\partial x}\left(\frac{\partial F_z}{\partial y} - \frac{\partial F_y}{\partial z}\right)
+
\frac{\partial}{\partial y}\left(\frac{\partial F_x}{\partial z} - \frac{\partial F_z}{\partial x}\right)
+
\frac{\partial}{\partial z}\left(\frac{\partial F_y}{\partial x} - \frac{\partial F_x}{\partial y}\right)
$$

$$
=
\underbrace{\frac{\partial^2 F_z}{\partial x \partial y}
-
\frac{\partial^2 F_z}{\partial y \partial x}}_{=0}
\;+\;
\underbrace{\frac{\partial^2 F_x}{\partial y \partial z}
-
\frac{\partial^2 F_x}{\partial z \partial y}}_{=0}
\;+\;
\underbrace{\frac{\partial^2 F_y}{\partial z \partial x}
-
\frac{\partial^2 F_y}{\partial x \partial z}}_{=0}
= 0
$$

* Curl describes local circulation
* Divergence measures sources or sinks  
* Consider a small volume:  
  $\Rightarrow$ inflow and outflow balance exactly  
  $\Rightarrow$  **Rotation alone cannot create a net outflow**
