## Electric dipoles

* two opposite charges separated by a distance
* electrically neutral overall
* electric fields do not fully cancel
* field pattern changes as charges move closer together
* dipoles are fundamental building blocks in matter

## Dipoles in nature & materials

* asymmetric charge distribution $\rightarrow$ electric dipoles can form
* examples:
    * water molecules
    * crystals & semiconductors
* dipoles can align in external electric fields
* important for dielectrics, biology, and piezoelectricity

## Electric dipole moment

* dipole strength depends on:
  * charge magnitude $q$
  * separation distance $\vec{\bf{d}}$
* define the electric dipole moment:

$$
\vec{\bf{p}} = q \vec{\bf{d}}
$$

* direction:
  * from negative to positive charge

## Electric dipole in electric field

* if placed in an electric field, a dipole experiences a torque
* the torque rotates the dipole toward the field direction (with $\vec{\bf{p}} = q \vec{\bf{d}}$)
$$
\vec{\bf{\tau}}=\vec{\bf{d}} \times \vec{\bf{F}}=\left(\frac{\vec{\bf{d}}}{2}\times q \vec{\bf{E}}\right)+\left(-\frac{\vec{\bf{d}}}{2}\times(-q \vec{\bf{E}})\right) =\vec{\bf{p}} \times \vec{\bf{E}}
$$
* aligned dipoles have lower potential energy (with $V = \frac{W}{q}$, $V = \vec{\bf{E}} \vec{\bf{d}}$, and $\vec{\bf{p}} = q \vec{\bf{d}}$)
$$
W = q V = \frac{\vec{\bf{p}}}{\vec{\bf{d}}} \; \vec{\bf{E}} \; \vec{\bf{d}} \quad \Rightarrow \quad U=-\vec{\bf{p}}\cdot \vec{\bf{E}}
$$

## How piezoelectric materials work

* asymmetric crystal structure e.g. due to dipoles $\rightarrow$ separated charge centers
* without stress: positive & negative centers cancel macroscopically
* mechanical stress shifts the charge distribution $\rightarrow$ **macroscopic polarization**
* effect is reversible: electric field can also deform the material
* electric polarization $\vec{\bf{P}} \propto \text{mechanical stress}$
* example: piezoelectric lighter generates a spark when pressed

## Inverse piezoelectric effect

* external electric fields shift charge centers inside the crystal $\Rightarrow$ crystal expands or contracts mechanically
* deformation depends on the field direction & material $\Delta l \propto E$
* mechanical oscillations used in speakers, buzzers, ultrasound transducers, etc.

## Quartz clocks

* a quartz crystal acts as a mechanical resonator
* due to piezoelectricity:
  * mechanical vibrations generate voltage
  * applied voltage causes vibrations
* electronics amplify and feed the signal back into the crystal
* the feedback loop plus battery compensates for energy losses and sustains oscillation
* quartz oscillates at a highly stable resonance frequency: $f = 32768\,\text{Hz} = 2^{15}\,\text{Hz}$ (above audible range & power of two)

## Wheatstone bridge

`el09`
* **goal:** determine unkown $R_x$
* **approach:** clever use of potentiometer/variable resistor, 2 known resistors and a voltmeter
* **idea:** balanced bridge $I_G = 0$: no current flows through the voltmeter $\rightarrow$ ratio gives unknown resistance
1. if $I_G = 0$ $\rightarrow$ **potential** at $B$ and $D$ same ($V_B = V_D$)
2. using Ohm's law and the ratio of equal voltage drops, i.e. $V_{DC} = V_{BC}, \; \& \; V_{AD} = V_{AB}$:
$$
\frac{V_{DC}}{V_{AD}}
=
\frac{V_{BC}}{V_{AB}}
\Rightarrow
\frac{I_2 R_2}{I_1 R_1}
=
\frac{I_x R_x}{I_3 R_3}
$$
4. with $I_1 = I_2$ and $I_3 = I_x$ (because no current between $D$ and $B$, i.e. $I_G=0$):
$$
R_x
=
\frac{R_2}{R_1}
R_3
$$

## Applications of Wheatstone bridges

* precise measurement of small resistance changes
* highly sensitive to bridge imbalance
* widely used with resistive sensors
* examples:
  * strain gauges
  * thermistors
  * pressure and force sensors
* measurable voltage appears when the bridge becomes unbalanced
* can be extended to measure capacitance or inductance

## Magnetic dipoles

* magnetic analogue of electric dipoles
* isolated magnetic monopoles have never been observed $\rightarrow$ **every magnet is dipole**
* moving electric charges generate magnetic dipole (magnetic field lines form closed loops)
* spin-$\frac{1}{2}$ particles such as electrons and protons posses magnetic moment

## Magnetic dipole in magnetic field
* assume a rectangular current loop with width $w$ and height $h$ in a uniform magnetic field
* loop is positioned such that the two wire segments experiencing force are perpendicular to $\vec{\bf{B}}$
* opposite sides experience opposite forces $\rightarrow$ net force cancels, **but torque remains**
$$\tau = F \frac{w}{2} + F \frac{w}{2} = Fw$$
* using magnetic Lorentz force on a wire segment $\vec{\bf{F}} = I \vec{\bf{h}} \times \vec{\bf{B}} = IhB $ and loop area $A = hw$:
$$\tau = IhwB = IAB $$
* define the magnetic dipole moment $\vec{\bf{\mu}} = I \vec{\bf{A}}$, we obtain the general solution of:
$$\vec{\bf{\tau}} = \vec{\bf{\mu}} \times \vec{\bf{B}}$$

## Spin and magnetic moments

* many particles possess an intrinsic angular momentum called **spin**
* spin makes particles behave like tiny gyroscopes
* charged particles with spin can possess magnetic dipole moments $\vec{\bf{\mu}} = I \vec{\bf{A}}$ 
* electrons and protons therefore behave like tiny magnets
* magnetic moments interact with external magnetic fields:
  * torque can align spins $\vec{\bf{\tau}} = \vec{\bf{\mu}} \times \vec{\bf{B}}$
  * moving charges can be deflected by magnetic Lorentz force
* basis for:
  * magnetism in materials
  * electron spin resonance
  * nuclear magnetic resonance (NMR) and MRI

## Force between parallel wires: Theory

* currents generate magnetic fields and magnetic fields exert forces on currents
* magnetic field of wire 1 at distance $d$:
$$
B_1
=
\frac{\mu_0}{2\pi}
\frac{I_1}{d}
$$
* force on wire 2 with current $I_2$:
$$
F_{21}
=
I_2 l B_1
=
\frac{\mu_0}{2\pi}
\frac{I_1 I_2}{d}
l
$$
* by Newton's 3rd axiom (action = reaction):
$$
F_{12} = -F_{21}
$$


## Ampère balance
`em04` 

* magnetic force on a current-carrying wire balanced against gravity
$$
F_{\mathrm{mag}} = F_g \quad \rightarrow \quad IlB = mg
$$

* magnetic force increases with current $I$ & wire length $l$ 
$$
F \propto I
\qquad
F \propto l
$$

* magnetic flux density remains constant for fixed magnetic field:
$$
B = \frac{F}{Il} = \textrm{const}
$$

## Moving-coil meter (Drehspulmesswerk)
`em41`

* current through a coil in magnetic field generates magnetic torque
* spiral spring provides restoring torque
* pointer rotates until both torques balance:
$$
M_{\mathrm{mag}}
=
M_{\mathrm{spring}}
\quad \rightarrow \quad
n I A B
=
D \alpha
\quad \rightarrow \quad
\alpha
=
\frac{nAB}{D} I
$$
* basis for analog ammeters, analog voltmeters, etc.
