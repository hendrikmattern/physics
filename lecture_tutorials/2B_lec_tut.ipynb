## Electrostatics vs. gravitation: Overview

Both forces follow an inverse-square law:

$$
F_e = k\frac{q_1 q_2}{r^2}
\qquad\qquad
F_g = G\frac{m_1 m_2}{r^2}
$$

with Coulomb's constant $k$ & gravitational constant  $G$

* Both act along the line connecting the two objects
* Coulomb force depends on **charge**, gravitational force on **mass**
* Coulomb force can be **attractive or repulsive**
* Gravitational force is **always attractive**


## Electrostatics vs. gravitation: Electron–proton interaction

Compare the relative strength of the Coulomb force and gravitational force between an electron and a proton

**Given:**

$q_e = -e$, $q_p = e$  
$m_e = 9.109 \times 10^{-31}\,\mathrm{kg}$  
$m_p = 1.673 \times 10^{-27}\,\mathrm{kg}$  
$k = 8.988 \times 10^9$  
$G = 6.674 \times 10^{-11}$

**What force do you think is stronger?**




## Electrostatics vs. gravitation: Electron–proton interaction (cont')
$$
\frac{|F_e|}{|F_g|} = \frac{k e^2}{G m_e m_p} \approx 2.264 \times 10^{39}
$$


* **$\rightarrow$ The electromagnetic force is enormously stronger than gravity on particle scales**
* If the particles are $1\,\mu\text{m}$ apart gravitationally, the Coulomb force would match that tiny force only at
$$
r \approx 4.762 \times 10^{13}\,\text{m} \approx 47,620 \, \text{million km}
$$
* This is far beyond planetary scales (distance between pluto and sun is approximately $5,906 \, \textrm{million km}$)


## Electrostatics vs. gravitation: Earth–Moon thought experiment

Determine the magnitude of the charge, $Q$, that must be placed on both the Earth and the Moon to make the magnitude of the electrostatic force between them equal to the gravitational force. 
Assume that the Earth and Moon can be treated as point charges located at their respective centers, and that they acquire equal charge.

**Given:**

Earth's mass: $m_E = 5.972 \times 10^{24} \, \text{kg}$  
Moon's mass: $m_M = 7.348 \times 10^{22} \, \text{kg}$  
average Earth-Moon distance $ r = 3.844 \times 10^8 \, \text{m}$  
$k = 8.988 \times 10^9 \, \text{Nm}^2/\text{C}^2$  
$G = 6.674 \times 10^{-11} \, \text{Nm}^2/\text{kg}^2$  



## Electrostatics vs. gravitation: Earth–Moon thought experiment (cont')

Set electrostatic and gravitational forces equal:

$$
G\frac{m_E m_M}{r^2} = k\frac{Q^2}{r^2}
$$

Solve for the equal charge on Earth and Moon:

$$
Q = \sqrt{\frac{G m_E m_M}{k}} \approx 5.714 \times 10^{13}\,\text{C}
$$


* Both the Earth and the Moon would need to have charges of approximately $5.714 \times 10^{13} \, \text{C} $ to exert the same magnitude of electrostatic and gravitational force.
* An absurd amount of charge, i.e. would require $3.567 \times 10^32$ elementary charges $e$ (number of stars in the observable universe is estimated to be around $10^{24}$)
* Large astronomical bodies are usually nearly electrically neutral

**$\rightarrow$ Gravity dominates on large scales**



## Electrostatics vs Gravitation: Final takeaway

- Both forces follow an inverse-square law: strength ∝ $1/r^2$  

$$
G\, \frac{m_1 m_2}{r^2} \quad \leftrightarrow \quad k\, \frac{q_1 q_2}{r^2}
$$

- Analogy: mass ↔ charge, gravitational constant $G$ ↔ Coulomb constant $k$  
- Key differences: 
    - $k \gg G$ → electrostatic force is vastly stronger → Coulomb force ≈ $10^{39}$ × gravity    
    - Gravity: always attractive; electrostatics: attractive **and** repulsive  
- Microscopic world: electrostatics dominates
- Macroscopic world: gravity dominates (astronomical bodies approx. electrically neutral)

## Oil droplet hovering in an electric field

If the droplet levitates, electric and gravitational forces balance (neglect buoyancy & drag):

$$
F_g = F_e
\qquad\Rightarrow\qquad
mg = qE
$$

With $E = V/d$ and $m = \rho_o \frac{4}{3}\pi r^3$:

$$
q = \frac{4\pi r^3 \rho_o g d}{3V}
$$



## Robert A. Millikan: Significance & Controversy

- **Key contribution**: 
  - oil-drop experiment → precise value of elementary charge $e$  
  - Established **quantization of charge** → milestone in atomic physics  
  - Nobel Prize in Physics (1923)  
- **Scientific controversy**:
  - did **not publish all measurements** (selected ~28 fitting data points)  
  - omission reduced scatter (critic Felix Ehrenhaft obtained highly variable results)  
- **Context & debate**:
  - Ehrenhaft suggested **non-quantized charge** (now known to be incorrect)  
  - historians partly **criticize**, others **defend** Millikan’s methods  
- **Additional aspect**:
  - contribution of Harvey Fletcher not credited (mutual agreement for publications; Flechter sole author on another paper as the basis of his PhD)  
- **Takeaway**:
  - result is **correct and foundational**  
  - highlights (challenges in) **data selection, credit, and scientific ethics**


## Application of Gauss’s Law to Non-conducting Uniformly Charged Sphere

- Goal: find $\mathbf{E}(r)$ using Gauss’s law  
- Sphere: 
    - radius $r_0$
    - total charge $Q$ (uniformly distributed)  
- Consider two regions:  
  - outside ($r > r_0$)  
  - inside ($r < r_0$)  

$$
\Phi = \oint_S \mathbf{E} \cdot d\mathbf{A} = \frac{Q_{\text{enc}}}{\varepsilon_0}
$$



## Non-conducting Uniformly Charged Sphere: $E$ outside the Sphere ($r > r_0$)

- Choose spherical Gaussian surface (symmetry)  
- Enclosed charge: total charge $Q$  
- $\vec{E}$ radial $\rightarrow$ constant on surface & perpendicular to $d\vec{A}$  
- surface of sphere is $4\pi r^2$

$$
\oint_S \mathbf{E} \cdot d\mathbf{A} = \frac{Q_{\text{enc}}}{\varepsilon_0}
$$

$$
E (4\pi r^2) = \frac{Q}{\varepsilon_0}
$$

$$
E = \frac{Q}{4\pi \varepsilon_0 r^2}
$$

**$\rightarrow$ Same as point charge**  



## Non-conducting Uniformly Charged Sphere: $E$ inside the Sphere ($r < r_0$)

- Uniform charge density with sphere volume $V =\frac{4}{3}\pi r_0^3$:
$$
\rho = \frac{Q}{V} = \frac{Q}{\frac{4}{3}\pi r_0^3}
$$
- Apply Gauss with enclosed charge $Q_{\text{enc}} = Q \frac{r^3}{r_0^3}$
$$
\oint_S \mathbf{E} \cdot d\mathbf{A} = \frac{Q_{\text{enc}}}{\varepsilon_0}
$$
$$
E (4\pi r^2) = \frac{Q r^3}{\varepsilon_0 r_0^3}
$$
$$
E = \frac{Q}{4\pi \varepsilon_0 r_0^3} \, r
$$
- Field increases **linearly with $r$**


## Non-conducting Uniformly Charged Sphere: Potential outside

- Use $V(\infty)=0$
$$
V(r) = -\int_{\infty}^{r} \vec{E}\cdot d\vec{l}
$$


- **Outside** ($r > r_0$):
$$
V(r) = -\int_{\infty}^{r} \frac{1}{4\pi\varepsilon_0}\frac{Q}{r'^2}\,dr'
$$
$$
V(r) = -\frac{Q}{4\pi\varepsilon_0}\int_{\infty}^{r} r'^{-2}\,dr'
$$
$$
V(r) = -\frac{Q}{4\pi\varepsilon_0}\left[-\frac{1}{r'}\right]_{\infty}^{r}
= \frac{1}{4\pi\varepsilon_0}\frac{Q}{r}
$$



## Non-conducting Uniformly Charged Sphere: Potential inside
- **Inside** ($r < r_0$): split path into outside + inside part
$$
V(r) = -\int_{\infty}^{r_0} \vec{E}\cdot d\vec{l} - \int_{r_0}^{r} \vec{E}\cdot d\vec{l}
$$
$$
V(r) = \frac{1}{4\pi\varepsilon_0}\frac{Q}{r_0}
-\int_{r_0}^{r} \frac{Q}{4\pi\varepsilon_0 r_0^3} r'\,dr'
$$
$$
V(r) = \frac{1}{4\pi\varepsilon_0}\frac{Q}{r_0}
-\frac{Q}{4\pi\varepsilon_0 r_0^3}\left[\frac{r'^2}{2}\right]_{r_0}^{r}
= \frac{1}{4\pi\varepsilon_0}\frac{Q}{r_0}
-\frac{Q}{4\pi\varepsilon_0 r_0^3}\frac{1}{2}(r^2-r_0^2)
$$
$$
V(r)=\frac{1}{4\pi\varepsilon_0}\frac{Q}{2r_0}\left(3-\frac{r^2}{r_0^2}\right)
$$
- Key idea:
  - outside → behaves like point charge  
  - inside → $E \propto r$, $V$ quadratic


## Electric Field & Potential of a Conducting Sphere

- All charge resides on **surface** (electrostatic equilibrium) 
- **Outside ($r > r_0$)** with $V(\infty)=0$:
$$
E(r)=\frac{1}{4\pi\varepsilon_0}\frac{Q}{r^2}
$$
$$
V(r)=\frac{1}{4\pi\varepsilon_0}\frac{Q}{r}
$$
- **Inside ($r < r_0$)** with $V(\infty)=0$:
$$
E(r)=0
$$
$$
V(r) = -\int_{\infty}^{r_0} \vec{E}\cdot d\vec{l} - \int_{r_0}^{r} \vec{E}\cdot d\vec{l} = V(r_0) - \int_{r_0}^{r} \vec{E}\cdot d\vec{l}
= \frac{1}{4\pi\varepsilon_0}\frac{Q}{r_0}
$$

- Key idea:
  - inside → **no field, constant potential**  
  - outside → behaves like **point charge**  


## Electric Field & Potential of a Hollow Sphere (Thin Shell)

- Spherical symmetry → choose spherical Gaussian surface  
- $\vec{E}$ radial, constant on surface, perpendicular to $d\vec{A}$  

$$
\oint_S \vec{E}\cdot d\vec{A} = E \cdot 4\pi r^2 = \frac{Q_{\text{enc}}}{\varepsilon_0}
$$

- **Electric field inside ($r < r_0$) is zero** (no enclosed charge: $Q_{\text{enc}}=0$):
$$
E(r)=0
$$
- **Electric field outside ($r > r_0$)** with enclosed charge: $Q_{\text{enc}}=Q$:
$$
E(r)=\frac{1}{4\pi\varepsilon_0}\frac{Q}{r^2}
$$

- Key idea:
    - **same results for $E$ & $V$ as for conducting sphere because same charge distribution**
    - inside → **no enclosed charge ⇒ no field**  
    - outside → behaves like **point charge**






## Electric Field of an Infinite Charged Plate

- Charge is distributed over a surface → define **surface charge density**, $\sigma$: charge per unit area (C/m²):
$$
\sigma = \frac{Q}{A}
$$
- Planar symmetry → $\vec{E}$ perpendicular to plate, same magnitude everywhere  
- Choose a **cylindrical Gaussian surface**:
  - flat faces parallel to plate  
  - centered on plate (one face above, one below)  
- Flux only through the two faces (no flux through side) with enclosed charge $Q_{\text{enc}} = \sigma A$
$$
\oint \vec{E}\cdot d\vec{A} = EA + EA = 2EA  = \frac{\sigma A}{\varepsilon_0} = \frac{Q_{\text{enc}}}{\varepsilon_0}
$$
$$
E=\frac{\sigma}{2\varepsilon_0}
$$
- Key idea:
  - field is **constant** (independent of distance)  
  - depends only on **surface charge density $\sigma$**

## Electric Potential of an Infinite Charged Plate

- Start from:
$$
V_B - V_A = -\int_A^B \vec{E}\cdot d\vec{l}
$$
- Uniform field → $\vec{E}\cdot d\vec{l}=E\,dx$
$$
V_B - V_A = -E(x_B - x_A)
$$
- Insert $E = \frac{\sigma}{2\varepsilon_0}$:
$$
V_B - V_A = -\frac{\sigma}{2\varepsilon_0}(x_B - x_A)
$$
- Choose reference (e.g. $V(0)=0$):
$$
V(x) = -\frac{\sigma}{2\varepsilon_0}x
$$
- Key idea:
  - potential changes **linearly** with distance  
  - only **potential differences** are physically relevant  


## Electric Field of Two Parallel Plates

- Two plates with surface charge densities $+\sigma$ and $-\sigma$  
- Each plate creates field $E=\frac{\sigma}{2\varepsilon_0}$  
- **Superposition**:
  - between plates → fields **add**
  - outside → fields **cancel**
- Resulting field:
$$
E = \frac{\sigma}{\varepsilon_0} \quad \text{(between plates)}
$$
$$
E = 0 \quad \text{(outside)}
$$
- Key idea:
  - field is **uniform between plates**  
  - nearly **zero outside** (ideal infinite plates)


## Electric Potential of Two Parallel Plates

- Start from:
$$
V_B - V_A = -\int_A^B \vec{E}\cdot d\vec{l}
$$
- Uniform field between plates → $\vec{E}\cdot d\vec{l}=E\,dx$
$$
V_B - V_A = -E(x_B - x_A)
$$
- Insert $E = \frac{\sigma}{\varepsilon_0}$:
$$
V_B - V_A = -\frac{\sigma}{\varepsilon_0}(x_B - x_A)
$$
- For plate separation $d$:
$$
V = E d = \frac{\sigma}{\varepsilon_0} d
$$
- Key idea:
  - potential changes **linearly between plates**  
  - defines a **constant voltage difference**
