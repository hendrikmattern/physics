James Clerk Maxwell unified electricity and magnetism through four fundamental equations, known as Maxwell’s equations. These equations describe how electric and magnetic fields interact and predict the existence of electromagnetic waves, which propagate through space at the speed of light.




## 2.8.X. Changing Electric Fields Produce Magnetic Fields; Displacement Current
One of Maxwell’s key contributions was the addition of the *displacement current* term to Ampère’s law. This term accounts for the fact that a changing electric field can generate a magnetic field, even in the absence of physical current.

- *Ampère’s Law (without Maxwell’s correction):*
  $$
\nabla \times \bf{B} = \mu_0 \bf{J} $$
  This states that a magnetic field is generated by an electric current.

- *With Displacement Current:*
  $$
\nabla \times \bf{B} = \mu_0 \left( \bf{J} + \epsilon_0 \frac{\partial \bf{E}}{\partial t} \right) $$
  The additional term \( \epsilon_0 \frac{\partial \bf{E}}{\partial t} \) accounts for the effect of a changing electric field producing a magnetic field, even in free space.

## 2.8.X. Gauss's Law for Magnetism
Gauss’s law for magnetism states that there are no magnetic monopoles; in other words, magnetic field lines always form closed loops.

$$
\nabla \cdot \bf{B} = 0 $$

This means that every magnetic field has both a north and south pole, unlike electric charges, which can exist separately.

## 2.8.X. Maxwell’s Equations
The four Maxwell’s equations in differential form are:

1. *Gauss's Law for Electricity:*
   $$
   
\nabla \cdot \bf{E} = \frac{\rho}{\epsilon_0} $$
   This states that electric charges produce electric fields, and the total flux of \( \bf{E} \) through a closed surface is proportional to the charge enclosed.

2. *Gauss’s Law for Magnetism:*
   $$
   
\nabla \cdot \bf{B} = 0 $$
   This implies that magnetic field lines have no beginning or end, meaning no magnetic monopoles exist.

3. *Faraday’s Law of Induction:*
   $$
   
\nabla \times \bf{E} = -\frac{\partial \bf{B}}{\partial t} $$
   This describes how a changing magnetic field induces an electric field, which is the principle behind electrical generators.

4. *Ampère-Maxwell Law:*
   $$
   
\nabla \times \bf{B} = \mu_0 \left( \bf{J} + \epsilon_0 \frac{\partial \bf{E}}{\partial t} \right) $$
   This extends Ampère’s Law by including the displacement current, showing that a changing electric field generates a magnetic field.

## 2.8.X. Production of Electromagnetic Waves
Maxwell’s equations predict that a time-varying electric field produces a magnetic field and vice versa. This self-sustaining process leads to the propagation of *electromagnetic waves* in space.

By taking the curl of Faraday’s Law and Ampère’s Law, we obtain wave equations for both \( \bf{E} \) and \( \bf{B} \):

$$
\nabla^2 \bf{E} = \mu_0 \epsilon_0 \frac{\partial^2 \bf{E}}{\partial t^2} $$
$$
\nabla^2 \bf{B} = \mu_0 \epsilon_0 \frac{\partial^2 \bf{B}}{\partial t^2} $$

These equations describe waves propagating at the speed \( c \), given by:

$$
c = \frac{1}{\sqrt{\mu_0 \epsilon_0}} $$

## 2.8.X. Electromagnetic Waves and Their Speed
From Maxwell’s equations, the speed of an electromagnetic wave in vacuum is:

$$
c = \frac{1}{\sqrt{\mu_0 \epsilon_0}} \approx 3.00 \times 10^8 \text{ m/s} $$

Electromagnetic waves are *transverse waves*, meaning the electric field \( \bf{E} \) and the magnetic field \( \bf{B} \) oscillate perpendicular to each other and to the direction of wave propagation.

## 2.8.X. Light as an Electromagnetic Wave and the Electromagnetic Spectrum
Light is a form of electromagnetic radiation. The electromagnetic spectrum includes:

- *Radio waves* (lowest frequency, longest wavelength)
- *Microwaves*
- *Infrared radiation*
- *Visible light*
- *Ultraviolet radiation*
- *X-rays*
- *Gamma rays* (highest frequency, shortest wavelength)

## 2.8.X. Measuring the Speed of Light
The speed of light was first accurately measured by *Ole Rømer* in the 17th century using the moons of Jupiter. Later, *Michelson* refined these measurements using rotating mirrors.

Modern methods involve laser interferometry and time-of-flight measurements.

## 2.8.X. Energy in Electromagnetic Waves and the Poynting Vector
Electromagnetic waves carry energy, which is described by the *Poynting vector*:

$$
\bf{S} = \frac{1}{\mu_0} \bf{E} \times \bf{B} $$

The magnitude of \( \bf{S} \) gives the power per unit area carried by the wave.

The *energy density* in an electromagnetic wave is:

$$
u = \frac{1}{2} \epsilon_0 E^2 + \frac{1}{2 \mu_0} B^2 $$

## 2.8.X. Radiation Pressure
Electromagnetic waves exert *radiation pressure* on surfaces. This is given by:

- *For a perfect absorber:*
  $$
P = \frac{S}{c} $$
- *For a perfect reflector:*
  $$
P = \frac{2S}{c} $$

This principle is used in applications like *solar sails* for spacecraft propulsion.

Welcome to this lecture script on Maxwell's Equations! As future engineers, understanding electromagnetism is crucial, and at the heart of this understanding lie Maxwell's Equations. Before Maxwell, electricity and magnetism were seen as separate phenomena.  However, through brilliant theoretical work and building upon experimental findings, James Clerk Maxwell unified these forces, revealing their deep interconnectedness. His equations not only summarized all known laws of electromagnetism at the time but also predicted new phenomena, most notably electromagnetic waves.

## 1. Gauss's Law for Electricity

### Equation:

**Integral Form:**
$$ \oint_S \vec{\bf{E}} \cdot d\vec{\bf{A}} = \frac{Q_{enc}}{\epsilon_0} $$

**Differential Form:**
$$ \nabla \cdot \vec{\bf{E}} = \frac{\rho}{\epsilon_0} $$

### Explanation:

Gauss's Law for Electricity describes the relationship between electric fields and electric charges. It states that the **electric flux** through any closed surface is proportional to the **enclosed electric charge**.

*   **Integral Form Explained:**
    *   $\oint_S$ : This integral symbol with a circle indicates integration over a closed surface *S*.
    *   $\vec{\bf{E}}$ :  This is the electric field vector. Remember, the electric field is a vector field that exerts a force on charged particles.
    *   $d\vec{\bf{A}}$ : This is a vector representing an infinitesimal area element on the closed surface *S*. Its magnitude *dA* is the area of the element, and its direction is normal to the surface, pointing outwards.
    *   $\vec{\bf{E}} \cdot d\vec{\bf{A}}$ : This is the dot product of the electric field and the area vector. It represents the component of the electric field that is perpendicular to the surface element, contributing to the flux through that element.
    *   $\oint_S \vec{\bf{E}} \cdot d\vec{\bf{A}}$ : This whole integral is the **electric flux** ($\Phi_E$) through the closed surface *S*. It's a measure of how much the electric field "flows" through the surface.
    *   $Q_{enc}$ : This is the total electric charge *enclosed* within the closed surface *S*.
    *   $\epsilon_0$ : This is the vacuum permittivity, a fundamental physical constant that determines the strength of electric interactions in a vacuum.

    In essence, the integral form says: "The total electric flux out of any closed surface is equal to the total charge enclosed by that surface, divided by $\epsilon_0$."

*   **Differential Form Explained:**
    *   $\nabla \cdot \vec{\bf{E}}$ : This is the divergence of the electric field.  The divergence of a vector field at a point measures how much the vector field "spreads out" or "diverges" from that point. In simpler terms, it's a measure of the source or sink of the field at that point.
    *   $\rho$ : This is the electric charge density at a point. It represents the amount of electric charge per unit volume at that point.

    The differential form says: "The divergence of the electric field at any point is proportional to the charge density at that point." This means that electric charges act as sources (for positive charges) or sinks (for negative charges) of the electric field, causing the field to diverge away from positive charges and converge towards negative charges.

### Derivation Idea:

Gauss's Law for Electricity is fundamentally derived from **Coulomb's Law**.  Imagine a single point charge. Coulomb's Law tells us the electric field it produces.  If you calculate the flux of this field through a spherical surface enclosing the charge, you'll find it's proportional to the charge.  Gauss's Law generalizes this for any charge distribution and any closed surface. It essentially re-expresses Coulomb's Law in terms of flux and enclosed charge, making it very powerful for calculating electric fields, especially in situations with symmetry.

## 2. Gauss's Law for Magnetism

### Equation:

**Integral Form:**
$$ \oint_S \vec{\bf{B}} \cdot d\vec{\bf{A}} = 0 $$

**Differential Form:**
$$ \nabla \cdot \vec{\bf{B}} = 0 $$

### Explanation:

Gauss's Law for Magnetism states that the **magnetic flux** through any closed surface is always **zero**. This is a direct consequence of the experimental observation that **magnetic monopoles (isolated north or south poles) do not exist**.  Magnetic field lines always form closed loops.

*   **Integral Form Explained:**
    *   $\oint_S$ : Integral over a closed surface *S*.
    *   $\vec{\bf{B}}$ : This is the magnetic field vector.
    *   $d\vec{\bf{A}}$ :  Infinitesimal area vector element on the closed surface *S*.
    *   $\vec{\bf{B}} \cdot d\vec{\bf{A}}$ : Dot product, representing the magnetic flux through the area element.
    *   $\oint_S \vec{\bf{B}} \cdot d\vec{\bf{A}}$ :  **Magnetic flux** ($\Phi_B$) through the closed surface *S*.

    The integral form says: "The total magnetic flux out of any closed surface is always zero."

*   **Differential Form Explained:**
    *   $\nabla \cdot \vec{\bf{B}}$ : Divergence of the magnetic field.

    The differential form says: "The divergence of the magnetic field is always zero everywhere."  This reinforces the idea that there are no magnetic monopoles.  Unlike electric fields which diverge from charges, magnetic fields always form closed loops, entering and leaving any closed surface in equal measure, thus resulting in zero net flux and zero divergence.

### Derivation Idea:

This law is fundamentally based on **experimental observation**.  Despite extensive searches, no isolated magnetic monopoles have ever been detected.  All magnetic sources are dipoles (like bar magnets), meaning they always have both a north and south pole. Because magnetic field lines are closed loops emanating from and returning to magnetic materials, for any closed surface, every magnetic field line that enters the surface must also exit it.  Hence, the net magnetic flux through any closed surface must be zero.

## 3. Faraday's Law of Induction

### Equation:

**Integral Form:**
$$ \oint_C \vec{\bf{E}} \cdot d\vec{\bf{l}} = - \frac{d}{dt} \iint_S \vec{\bf{B}} \cdot d\vec{\bf{A}} $$

**Differential Form:**
$$ \nabla \times \vec{\bf{E}} = - \frac{\partial \vec{\bf{B}}}{\partial t} $$

### Explanation:

Faraday's Law of Induction describes how a **changing magnetic field induces an electric field**. This principle is fundamental to the operation of generators, transformers, and many other electromagnetic devices.

*   **Integral Form Explained:**
    *   $\oint_C$ :  Integral around a closed loop *C*.
    *   $\vec{\bf{E}}$ : Electric field vector.
    *   $d\vec{\bf{l}}$ : Infinitesimal vector element along the closed loop *C*. Its direction is tangent to the loop.
    *   $\vec{\bf{E}} \cdot d\vec{\bf{l}}$ : Dot product.  $\oint_C \vec{\bf{E}} \cdot d\vec{\bf{l}}$ is the **electromotive force (EMF)** around the closed loop *C*.  EMF is the voltage generated around a closed loop due to the induced electric field.
    *   $- \frac{d}{dt} \iint_S$ :  Negative time derivative of the integral over a surface *S* bounded by the loop *C*.
    *   $\iint_S \vec{\bf{B}} \cdot d\vec{\bf{A}}$ : This is the **magnetic flux** ($\Phi_B$) through the surface *S* enclosed by the loop *C*.  *S* is any surface whose boundary is the loop *C*.

    The integral form says: "The electromotive force (EMF) induced around any closed loop is equal to the negative rate of change of magnetic flux through any surface bounded by that loop." The negative sign (Lenz's Law) indicates that the induced EMF opposes the change in magnetic flux that produces it.

*   **Differential Form Explained:**
    *   $\nabla \times \vec{\bf{E}}$ : Curl of the electric field. The curl of a vector field at a point measures the tendency of the field to "circulate" or "rotate" around that point.
    *   $- \frac{\partial \vec{\bf{B}}}{\partial t}$ :  Negative partial time derivative of the magnetic field.  $\frac{\partial \vec{\bf{B}}}{\partial t}$ represents how the magnetic field is changing with time.

    The differential form says: "The curl of the electric field at any point is equal to the negative rate of change of the magnetic field at that point." This means that a time-varying magnetic field creates a circulating electric field.

### Derivation Idea:

Faraday's Law is based on **experimental observations** made by Michael Faraday. He discovered that moving a magnet near a coil of wire, or changing the current in one coil near another, induces a voltage (EMF) in the coil. The magnitude of the induced EMF is proportional to the rate of change of magnetic flux.  The direction of the induced current (and thus the induced electric field) opposes the change in flux (Lenz's Law).  The mathematical formulation, particularly the relationship between EMF and the rate of change of flux, is a generalization and quantification of these experimental findings.  Using Stokes' Theorem, the integral form can be converted to the differential form.

## 4. Ampere-Maxwell's Law

### Equation:

**Integral Form:**
$$ \oint_C \vec{\bf{B}} \cdot d\vec{\bf{l}} = \mu_0 \left( I_{enc} + \epsilon_0 \frac{d}{dt} \iint_S \vec{\bf{E}} \cdot d\vec{\bf{A}} \right) $$

**Differential Form:**
$$ \nabla \times \vec{\bf{B}} = \mu_0 \left( \vec{\bf{J}} + \epsilon_0 \frac{\partial \vec{\bf{E}}}{\partial t} \right) $$

### Explanation:

Ampere-Maxwell's Law, often just called Ampere's Law with Maxwell's addition, describes how a **magnetic field is generated by both electric current and a changing electric field**. This is a crucial extension to the original Ampere's Law made by Maxwell, incorporating the concept of **displacement current**.

*   **Integral Form Explained:**
    *   $\oint_C$ :  Integral around a closed loop *C*.
    *   $\vec{\bf{B}}$ : Magnetic field vector.
    *   $d\vec{\bf{l}}$ : Infinitesimal vector element along the closed loop *C*.
    *   $\oint_C \vec{\bf{B}} \cdot d\vec{\bf{l}}$ :  This integral relates to the **magnetomotive force (MMF)**, analogous to EMF for electric fields, but for magnetic fields around a closed loop.
    *   $\mu_0$ : Vacuum permeability, a fundamental constant that determines the strength of magnetic interactions in a vacuum.
    *   $I_{enc}$ : This is the total **conduction current** (flow of charges) that passes through any surface *S* bounded by the loop *C*.
    *   $\epsilon_0 \frac{d}{dt} \iint_S \vec{\bf{E}} \cdot d\vec{\bf{A}}$ : This term is **Maxwell's displacement current**.
        *   $\iint_S \vec{\bf{E}} \cdot d\vec{\bf{A}}$ is the electric flux ($\Phi_E$) through the surface *S*.
        *   $\frac{d}{dt} \iint_S \vec{\bf{E}} \cdot d\vec{\bf{A}} = \frac{d\Phi_E}{dt}$ is the rate of change of electric flux.
        *   $\epsilon_0 \frac{d\Phi_E}{dt}$ is the displacement current, effectively acting as another source of magnetic field, just like conduction current.

    The integral form says: "The circulation of the magnetic field around any closed loop is proportional to the sum of the conduction current and the displacement current through any surface bounded by that loop."

*   **Differential Form Explained:**
    *   $\nabla \times \vec{\bf{B}}$ : Curl of the magnetic field.
    *   $\vec{\bf{J}}$ :  **Conduction current density** vector. It represents the current per unit area.
    *   $\epsilon_0 \frac{\partial \vec{\bf{E}}}{\partial t}$ : **Displacement current density**.  $\frac{\partial \vec{\bf{E}}}{\partial t}$ is the partial time derivative of the electric field.

    The differential form says: "The curl of the magnetic field at any point is proportional to the sum of the conduction current density and the displacement current density at that point." This shows that both electric current and a time-varying electric field create circulating magnetic fields.

### Derivation Idea:

The original **Ampere's Circuital Law** (without Maxwell's addition) only accounted for magnetic fields generated by conduction currents. However, when considering circuits with capacitors, Ampere's original law led to inconsistencies, particularly with charge conservation and in time-varying situations.

Maxwell realized that to make Ampere's Law consistent and to describe electromagnetic phenomena completely, he needed to add the **displacement current term**. He hypothesized that a changing electric field in space could also produce a magnetic field, just like a real current. This was a stroke of genius and a crucial step towards unifying electricity and magnetism.

The necessity of displacement current can be shown by considering charging capacitor. Between the capacitor plates there is no conduction current but there is a changing electric field as charge builds up on the plates. Maxwell argued that this changing electric field acts as a "displacement current", creating a magnetic field in the region between the plates, just as a conduction current would.  This addition resolved inconsistencies and paved the way for the prediction of electromagnetic waves.  Using Stokes' Theorem, the integral form can be related to the differential form.

## The Displacement Current in Detail

Maxwell's displacement current, $\epsilon_0 \frac{\partial \vec{\bf{E}}}{\partial t}$, is not a flow of actual charges like conduction current. Instead, it represents the effect of a **time-varying electric field** acting as a source of magnetic field.

**Why is it important?**

1.  **Consistency of Ampere's Law:**  Without displacement current, Ampere's Law would be inconsistent in certain situations, particularly with time-varying fields and circuits containing capacitors. Maxwell's addition makes Ampere's Law universally valid.

2.  **Electromagnetic Waves:** The inclusion of displacement current is essential for the prediction of electromagnetic waves.  Maxwell's equations, with the displacement current term, show that a changing electric field creates a magnetic field, and a changing magnetic field creates an electric field (Faraday's Law). This interplay between changing electric and magnetic fields allows for self-sustaining electromagnetic waves to propagate through space, even in vacuum. Light is an electromagnetic wave, and Maxwell's equations beautifully predicted its existence and properties.

3.  **Unification of Electromagnetism:**  Displacement current is a key element in the unification of electricity and magnetism. It highlights that changing electric fields are just as capable of generating magnetic fields as electric currents, further demonstrating the deep interdependence of these two phenomena.

## Maxwell's Equations in Vacuum and in Matter

### Maxwell's Equations in Vacuum:

In vacuum (free space, no charges or materials present), the equations simplify because charge density ($\rho$) and current density ($\vec{\bf{J}}$) are zero.

1.  **Gauss's Law for Electricity:**  $\nabla \cdot \vec{\bf{E}} = 0$
2.  **Gauss's Law for Magnetism:** $\nabla \cdot \vec{\bf{B}} = 0$
3.  **Faraday's Law of Induction:** $\nabla \times \vec{\bf{E}} = - \frac{\partial \vec{\bf{B}}}{\partial t}$
4.  **Ampere-Maxwell's Law:** $\nabla \times \vec{\bf{B}} = \mu_0 \epsilon_0 \frac{\partial \vec{\bf{E}}}{\partial t}$

These vacuum equations are incredibly important because they describe the behavior of electromagnetic waves, including light, in free space.  From these equations, one can derive the wave equation for both electric and magnetic fields, and calculate the speed of these waves, which turns out to be the speed of light!

### Maxwell's Equations in Matter:

In materials, we need to consider the effects of the material's response to electric and magnetic fields. This involves introducing macroscopic quantities:

*   **Electric Displacement Field ($\vec{\bf{D}}$):**  Accounts for the polarization of the material in response to an electric field.  $\vec{\bf{D}} = \epsilon_0 \vec{\bf{E}} + \vec{\bf{P}}$, where $\vec{\bf{P}}$ is the electric polarization.
*   **Magnetic Field Intensity ($\vec{\bf{H}}$):** Accounts for the magnetization of the material in response to a magnetic field. $\vec{\bf{H}} = \frac{1}{\mu_0} \vec{\bf{B}} - \vec{\bf{M}}$, where $\vec{\bf{M}}$ is the magnetization.

In linear, isotropic, and homogeneous (LIH) materials, we often have simple relationships:
$\vec{\bf{D}} = \epsilon \vec{\bf{E}} = \epsilon_r \epsilon_0 \vec{\bf{E}}$ and $\vec{\bf{B}} = \mu \vec{\bf{H}} = \mu_r \mu_0 \vec{\bf{H}}$, where $\epsilon_r$ is the relative permittivity and $\mu_r$ is the relative permeability of the material.

Maxwell's Equations in Matter (in differential form, using free charge and current densities):

1.  **Gauss's Law for Electricity in Matter:** $\nabla \cdot \vec{\bf{D}} = \rho_{free}$  (where $\rho_{free}$ is the free charge density)
2.  **Gauss's Law for Magnetism:** $\nabla \cdot \vec{\bf{B}} = 0$ (remains unchanged)
3.  **Faraday's Law of Induction:** $\nabla \times \vec{\bf{E}} = - \frac{\partial \vec{\bf{B}}}{\partial t}$ (remains unchanged)
4.  **Ampere-Maxwell's Law in Matter:** $\nabla \times \vec{\bf{H}} = \vec{\bf{J}}_{free} + \frac{\partial \vec{\bf{D}}}{\partial t}$ (where $\vec{\bf{J}}_{free}$ is the free current density)

In matter, these equations describe electromagnetism in the presence of materials, taking into account polarization and magnetization effects.

## Significance of Maxwell's Equations

Maxwell's Equations are much more than just a set of equations describing electricity and magnetism. Their significance is profound and far-reaching:

1.  **Unification of Electricity and Magnetism:** They revealed that electricity and magnetism are not separate forces, but rather different aspects of a single electromagnetic force. They are intrinsically linked, as demonstrated by Faraday's Law and Ampere-Maxwell's Law.

2.  **Prediction of Electromagnetic Waves and Light:**  Maxwell's equations predicted the existence of electromagnetic waves that propagate at a speed calculated from $\epsilon_0$ and $\mu_0$. This speed turned out to be the speed of light! This was a revolutionary discovery, showing that light itself is an electromagnetic wave. This unified optics with electromagnetism.

3.  **Foundation of Classical Electromagnetism:** Maxwell's equations form the complete and consistent foundation of classical electromagnetism. They describe all classical electromagnetic phenomena. They are used in countless applications across engineering and physics, from electrical circuits and antennas to motors, generators, and telecommunications.

4.  **Basis for Special Relativity:**  While formulated within classical physics, Maxwell's equations played a crucial role in the development of Einstein's theory of Special Relativity.  The fact that the speed of light is constant for all observers, a postulate of special relativity, is deeply connected to Maxwell's equations and the properties of electromagnetic waves.

## Summary of Maxwell's Equations

Here is a concise summary of Maxwell's Equations, both in integral and differential forms:

**Maxwell's Equations in Vacuum (Differential Form):**

1.  $\nabla \cdot \vec{\bf{E}} = 0$  (Gauss's Law for Electricity)
2.  $\nabla \cdot \vec{\bf{B}} = 0$  (Gauss's Law for Magnetism)
3.  $\nabla \times \vec{\bf{E}} = - \frac{\partial \vec{\bf{B}}}{\partial t}$ (Faraday's Law of Induction)
4.  $\nabla \times \vec{\bf{B}} = \mu_0 \epsilon_0 \frac{\partial \vec{\bf{E}}}{\partial t}$ (Ampere-Maxwell's Law)

**Maxwell's Equations in Vacuum (Integral Form):**

1.  $\oint_S \vec{\bf{E}} \cdot d\vec{\bf{A}} = \frac{Q_{enc}}{\epsilon_0} \xrightarrow{\rho=0} 0$ (Gauss's Law for Electricity in vacuum)
2.  $\oint_S \vec{\bf{B}} \cdot d\vec{\bf{A}} = 0$ (Gauss's Law for Magnetism)
3.  $\oint_C \vec{\bf{E}} \cdot d\vec{\bf{l}} = - \frac{d}{dt} \iint_S \vec{\bf{B}} \cdot d\vec{\bf{A}}$ (Faraday's Law of Induction)
4.  $\oint_C \vec{\bf{B}} \cdot d\vec{\bf{l}} = \mu_0 \left( I_{enc} + \epsilon_0 \frac{d}{dt} \iint_S \vec{\bf{E}} \cdot d\vec{\bf{A}} \right) \xrightarrow{I_{enc}=0} \mu_0 \epsilon_0 \frac{d}{dt} \iint_S \vec{\bf{E}} \cdot d\vec{\bf{A}}$ (Ampere-Maxwell's Law in vacuum, assuming no conduction current in the region of interest but changing E field possible)
